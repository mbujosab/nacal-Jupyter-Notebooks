## Distintos tipos de números y variables simbólicas



El módulo NAcAL no solo trabaja con números enteros, o de coma flotante, también con fracciones, e incluso con variables simbólicas, pues usa el módulo [SymPy]([https://www.sympy.org/en/index.html](https://www.sympy.org/en/index.html)). 
Por ejemplo



```python
a = sympy.symbols('a')               # Variable simbólica
1 + .02 + fracc(1,2) + a + sympy.pi  # entero +  coma flotante + racional + Variable simbólica + irracional
```

```python
x, y, z = sympy.symbols('x y z')
```

```python
x = Vector([x,y,z])
x
```

```python
a * x + sympy.sqrt(2)* x            # Vector con variables simbólicas
```

## Sistema de ecuaciones



```python
b1, b2, b3 = sympy.symbols('b_1 b_2 b_3')
```

```python
b = Vector([b1, b2, b3])
```

```python
A = Matrix([[1,2,2,2],[2,4,6,8],[3,6,8,10]])
A
```

```python
TrC = SEL(A,b, True).TrC
E = I(5) & TrC
E
```

#### Solución particular



```python
xp = slice(1,4) | E | 0   # los cuatro primeros números de la última columna
xp
```

```python
A * xp
```

### Dando valores



```python
b.subs([(b1,1), (b2,5), (b3,6)])
```

```python
SEL(A, b.subs([(b1,1), (b2,5), (b3,6)]), 1)
```

```python
SEL(A, b.subs([(b1,-sympy.sqrt(3)), (b2,sympy.pi), (b3,sympy.pi-sympy.sqrt(3))]), 1)
```

```python
xp = SEL(A, b.subs([(b1,-sympy.sqrt(3)), (b2,sympy.pi), (b3,sympy.pi-sympy.sqrt(3))]), 1).solP
xp
```

```python
A * xp
```

#### soluciones especiales



```python
A * Vector([-2,1,0,0])
```

```python
A * (slice(1,4) | E | 4)    # los paréntesis son importantes aquí
```

```python
K = A.copy() & TrC   # Forma pre-escalonada de A.

# Si K tiene columnas nulas, las correspondientes de E son soluciones a Ax=0

SolEspeciales = Sistema( [ slice(1,A.n)|E|j  for j,col in enumerate(K) if col.es_nulo() ] )

SolEspeciales
```
