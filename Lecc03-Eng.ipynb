#### A Bit of Python Before We Start



**Index Lists.**
The class `range(i,j)` returns an object that produces a sequence of integers from the start `i` (inclusive) to the end `j` (exclusive). Thus, `range(i, j)` produces `i, i+1, i+2, ..., j-1`. The default starting value is `0`. So, `range(4)` produces `0`, `1`, `2`, `3`. When a third argument is provided for step (`range(start, stop[, step])`), it specifies the increment (or decrement) between indices.



```python
list(range(5))
```

Remember that Python starts counting at `0` by default. Thus, the previous list (of 5 elements) starts at `0`.

Therefore, if we want a list of integers from 1 to 9, we will write



```python
list(range(1,10))
```

**Randomly Ordered Index Lists.**

With the method `random.sample(list, length)`, taking random values from the list `list` (without replacement, meaning no element from `list` is taken more than once; therefore, `length` cannot be a number larger than the number of elements in the list), we generate a list with the specified `length`.

#### Let's Replicate What Is Stated in Section 3.B of the Book



Let’s generate a list $\boldsymbol{\alpha}$ with `n` indices having values between `1` and `p`, where `p` is greater than or equal to $n$; for example:



```python
n = 4
p = 7
alpha = random.sample(range(1, p+1), n) # list of n random indices less than or equal to p
print(alpha)
```

Now, with the list of indices $\boldsymbol{\alpha}$, we will create a submatrix formed by `n` rows of the identity matrix $\Mat{I}$ of order $p$.



```python
alpha|I(n+3)
```

Remember that Python starts counting at `0`; therefore, the *first* number in the list of indices $\boldsymbol{\alpha}$ is:



```python
alpha[0]
```

It is easy to see that the *first* number in the list $\boldsymbol{\alpha}$ indicates where the *first* column of the identity matrix of order $n$ is located:



```python
print(alpha[0])  # first element of the index list
alpha|I(n+3)|(alpha[0])
```

That the *second* number in the list $\boldsymbol{\alpha}$ indicates where the *second* column of the identity matrix of order $n$ is located:



```python
print(alpha[1])  # second element of the index list
alpha|I(n+3)|(alpha[1])
```

That the *third* number in the list $\boldsymbol{\alpha}$ indicates where the *third* column of the identity matrix of order $n$ is located:



```python
print(alpha[2])  # third element of the index list
alpha|I(n+3)|(alpha[2])
```

etcetera&#x2026;

Furthermore, the columns that occupy positions not in the list $\boldsymbol{\alpha}$ are null. Therefore, in the example there are `p-n` null columns:



```python
print(p-n)
print(alpha)
display(alpha|I(n+3))
```

Thus, we can generate the corresponding system to the columns of the identity matrix of order `n` by selecting the columns of the submatrix `alpha|I(k)` following consecutively the indices of the list $\boldsymbol{\alpha}$ with `n` indices, and where `k` is <u>any number greater than or equal to `n`</u>:



```python
Sistema([ alpha|I(57)|j for j in alpha ])
```

If instead of selecting columns whose indices are in $\boldsymbol{\alpha}$, we use the indices that are not in $\boldsymbol{\alpha}$, we obtain null columns:



```python
Sistema([ alpha|I(57)|j for j in range(1,p+1) if j not in alpha ])
```

Let’s check what happens when multiplying 
$$
\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}
$$



```python
(I(p)|alpha)*(alpha|I(p))
```

On the diagonal, there are ones in the positions of the indices of $\boldsymbol{\alpha}$ and zeros elsewhere.

Consequently, if $\boldsymbol{\alpha}$ contains all indices from `1` to `p`, then, 
$$
\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}} = \Mat{I}.
$$ 

Therefore, imagine that we distribute the indices from `1` to `p`  (for example from `1` to `7`) into three sublists ($\boldsymbol{\gamma}_1$, $\boldsymbol{\gamma}_2$, $\boldsymbol{\gamma}_3$), so that each index is in one, and only one, of the three sublists. For example:



```python
gamma1 = (2,4,6,)
gamma2 = (1,7,)
gamma3 = (3,5,)
```

If we concatenate the lists $\boldsymbol{\gamma}_1$, $\boldsymbol{\gamma}_2$, and $\boldsymbol{\gamma}_3$ (*Python lists are concatenated with the operation* `+`), we will have a list $\boldsymbol{\alpha}$ containing all indices (although in arbitrary order):



```python
alpha = gamma1 + gamma2 + gamma3
alpha
```

That is
$$\boldsymbol{\alpha}=\boldsymbol{\gamma}_1\concat\boldsymbol{\gamma}_2\concat\boldsymbol{\gamma}_3$$
where $\concat$ indicates the concatenation of the lists.



##### Creating the Identity Matrix as a Sum of Matrices



Consequently, we have that the sum of the following matrices
$$
\elemRPE{\Mat{I}}{\gamma1}\elemLPE{\Mat{I}}{\gamma1} + \elemRPE{\Mat{I}}{\gamma2}\elemLPE{\Mat{I}}{\gamma2} + \elemRPE{\Mat{I}}{\gamma3}\elemLPE{\Mat{I}}{\gamma3} $$
must equal the identity matrix. Let’s see:



```python
(I(p)|gamma1)*(gamma1|I(p)) + (I(p)|gamma2)*(gamma2|I(p)) + (I(p)|gamma3)*(gamma3|I(p))
```

In case you do not see it, let's visualize the different summands using successive systems:



```python
for lista in [gamma1, gamma2, gamma3]:
    display(Sistema([lista, I(p)|lista, lista|I(p), (I(p)|lista)*(lista|I(p))]) )
```

&#x2026; the first element of each system is the list of indices $\boldsymbol{\gamma}_i$, the second is the corresponding selection of the columns of the identity matrix $\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_i}$, the third the corresponding selection of the rows of the identity matrix $\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_i}$ and the fourth the product matrix $\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_i}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_i}$. 

If you sum the last matrices from each system, you will see that you obtain the identity matrix $\Mat{I}$ of order 7.

\*Recap

With the above, we have that if $\boldsymbol{\alpha}$ is the concatenation of index lists $\boldsymbol{\gamma}_1$, $\boldsymbol{\gamma}_2,\ldots,\boldsymbol{\gamma}_k$ (for example, three):
$$
\boldsymbol{\alpha}=\boldsymbol{\gamma}_1\concat\boldsymbol{\gamma}_2\concat\boldsymbol{\gamma}_3,
$$
so that in the complete list $\boldsymbol{\alpha}$ are all indices from $1$ to $p$, with none repeated, then:
$$\Mat{I}= 
\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}=
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_1} + \elemRPE{\Mat{I}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_2} + \elemRPE{\Mat{I}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_3}$$

We will use these equalities, along with the properties of the matrix product, to find other ways to calculate matrix products.



##### Calculating the Matrix Product $\Mat{AB}$ with the Columns of $\Mat{A}$ and the Rows of $\Mat{B}$ Reordered



Since
$$\Mat{AB}=\Mat{AIB}=
\Mat{A}\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}\Mat{B}=
%\Big(\Mat{A}\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\Big)\Big(\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}\Mat{B}\Big)=
\elemRPE{\Mat{A}}{\boldsymbol{\alpha}}\elemLPE{\Mat{B}}{\boldsymbol{\alpha}}$$
we can arbitrarily reorder the columns of matrix $\Mat{A}$ and, with the same reordering, the rows of matrix $\Mat{B}$, so that the product of the two new matrices $\elemRPE{\Mat{A}}{\boldsymbol{\alpha}}\elemLPE{\Mat{B}}{\boldsymbol{\alpha}}$ is equal to the product $\Mat{AB}$.

Let’s generate two matrices $\Mat{A}$ of order $m$, $p$ and $\Mat{B}$ of order $p$, $n$;



```python
# For example 
m = 3
p = 7
A = Matrix( [ [ (1+i+j)  for i in range(p) ] for j in range(m) ] )

n = 2
B = Matrix( [ [ (1+j+i*j)  for i in range(n) ] for j in range(p) ] )
```

Reorder the `p` indices randomly:



```python
indices = random.sample(range(1, A.n+1), A.n)
print(indices)
```

Represent the original matrices



```python
Sistema([A,B])
```

Represent the reordered matrices



```python
Sistema([A|indices, indices|B])
```

And calculate both products: $\Mat{AB}$ on the left and $\elemRPE{\Mat{A}}{\boldsymbol{\alpha}}\elemLPE{\Mat{B}}{\boldsymbol{\alpha}}$ on the right&#x2026; they are equal!



```python
Sistema([A*B, (A|indices)*(indices|B),])
```

##### Calculating the Matrix Product as a Sum of Matrices



Since
$$\Mat{AB}=\Mat{AIB}=
\Mat{A}\Big(
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_1} + 
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_2} + 
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_3}
\Big)\Mat{B}=
\Mat{A}\Big(\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_1}\Big)\Mat{B} + \Mat{A}\Big(\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_2}\Big)\Mat{B} + \Mat{A}\Big(\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_3}\Big)\Mat{B}=
\elemRPE{\Mat{A}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{B}}{\boldsymbol{\gamma}_1} + 
\elemRPE{\Mat{A}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{B}}{\boldsymbol{\gamma}_2} + 
\elemRPE{\Mat{A}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{B}}{\boldsymbol{\gamma}_3} $$
we can calculate $\Mat{AB}$ using sums of products of submatrices.

Now consider two matrices $\Mat{A}$ of order $m$, $p$ and $\Mat{B}$ of order $p$, $n$;



```python
m = 3
p = 6
A = Matrix( [ [ (1+i+j)  for i in range(p) ] for j in range(m) ] )
A
```

```python
n = 2
B = Matrix( [ [ (2*(j-2*i))  for i in range(n) ] for j in range(p) ] )
B
```

Now we can consider even indices on one side, and odd indices on the other:



```python
evens   = [2, 4, 6]
odds    = [1, 3, 5]
```

Represent the selected submatrices with even indices, as well as their product:



```python
Sistema([A|evens, evens|B, (A|evens)*(evens|B)])
```

And do the same using the odd indices:



```python
Sistema([A|odds, odds|B, (A|odds)*(odds|B)])
```

Now let's calculate $\Mat{AB}$ on one side, and sum 
$
\elemRPE{\Mat{A}}{evens}\elemLPE{\Mat{B}}{evens} + \elemRPE{\Mat{A}}{odds}\elemLPE{\Mat{B}}{odds}
$



```python
Sistema([A*B, (A|evens)*(evens|B)+(A|odds)*(odds|B)])
```
