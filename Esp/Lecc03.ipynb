## Producto de matrices



En la [sección 3.1](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.3.1) del libro (y en el código del módulo [NAcAL](https://github.com/mbujosab/nacallib)), se define el producto de una matriz $\Mat{A}$ de orden $m\times p$ por otra matriz $\Mat{B}$ de orden $p\times n$ como aquella matriz cuya columna $j$-ésima es
$$
\elemRP{\Mat{AB}}{j}=\Mat{A}\elemRPE{\Mat{B}}{j}.
$$
Es decir, cada columna de $\Mat{AB}$ es una combinación lineal (es decir, matriz por vector) de las columnas de $\Mat{A}$.

Compruebe con ejemplos que efectivamente es así. Por ejemplo



```python
x = Vector((1, 1, -2)); y = Vector((1, -2, 1)); z = Vector((-2, 1, 1))
A = Matrix([x,y,z])
B = Matrix([z,x])
```

```python
(A*B)|2
```

```python
A*(B|2)
```

**Listas de índices.**

El valor inicial por defecto es `0`. Así, `range(4)` produce `0`, `1`, `2`, `3`. Cuando se proporciona un tercer argumento de paso (`range(start, stop[, paso])`), éste especifica el incremento (o decremento) entre índices.



```python
list(range(5))
```

Recuerde que Python empieza a contar por `0` por defecto. Por eso, la anterior lista (de 5 elementos) empieza por `0`.

Por tanto, si queremos una lista de enteros de 1 a 9, escribiremos



```python
list(range(1,10))
```

Generemos una lista $\boldsymbol{\alpha}$ con `n` índices con valores entre el `1` y `p`, donde `p` es mayor o igual a $n$; por ejemplo:



```python
n = 4
p = 7
alpha = random.sample(range(1, p+1), n) # lista de n índices al azar menores o iguales a p
print(alpha)
```

Ahora, con la lista de índices $\boldsymbol{\alpha}$, vamos a crear una submatriz formada por `n` de las filas de la matriz identidad $\Mat{I}$ de orden $p$.



```python
alpha|I(n+3)
```

Recuerde que Python comienza a contar por `0`; por tanto, el *primer* número de la lista de índices $\boldsymbol{\alpha}$ es:



```python
alpha[0]
```

Es fácil ver que el *primer* número de la lista $\boldsymbol{\alpha}$ indica dónde está la *primera* columna de la matriz identidad de orden $n$:



```python
print(alpha[0])  # primer elemento de la lista de índices
alpha|I(n+3)|(alpha[0])
```

Que el *segundo* número de la lista $\boldsymbol{\alpha}$ indica dónde está la *segunda* columna de la matriz identidad de orden $n$:



```python
print(alpha[1])  # segundo elemento de la lista de índices
alpha|I(n+3)|(alpha[1])
```

Que el *tercer* número de la lista $\boldsymbol{\alpha}$ indica dónde está la *tercera* columna de la matriz identidad de orden $n$:



```python
print(alpha[2])  # tercer elemento de la lista de índices
alpha|I(n+3)|(alpha[2])
```

etcétera&#x2026;

Además, son nulas las columnas que ocupan posiciones que no están en la lista $\boldsymbol{\alpha}$. Por tanto, en el ejemplo hay `p-n` columnas nulas:



```python
print(p-n)
print(alpha)
display(alpha|I(n+3))
```

Así, podemos generar el sistema correspondiente a las columnas de la matriz identidad de orden `n` seleccionando las columnas de la submatriz `alpha|I(k)` siguiendo consecutivamente los índices de la lista $\boldsymbol{\alpha}$ con `n` índices, y donde `k` es <u>cualquier número mayor o igual que `n`</u>:



```python
Sistema([ alpha|I(57)|j for j in alpha ])
```

Si en lugar de seleccionar columnas cuyos índices están en $\boldsymbol{\alpha}$, usamos los índices que no están en $\boldsymbol{\alpha}$, obtenemos columnas nulas



```python
Sistema([ alpha|I(57)|j for j in range(1,p+1) if j not in alpha ])
```

Comprobemos lo que pasa al multiplicar 
$$
\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}
$$



```python
(I(p)|alpha)*(alpha|I(p))
```

En la diagonal hay unos en las posiciones de los índices de $\boldsymbol{\alpha}$ y ceros en el resto.

En consecuencia, si $\boldsymbol{\alpha}$ contiene todos los índices desde `1` a `p`, entonces, 
$$
\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}} = \Mat{I}.
$$ 

Por tanto, imagine que distribuimos los índices de `1` a `p`  (por ejemplo de `1` a `7`) en tres sublistas ($\boldsymbol{\gamma}_1$, $\boldsymbol{\gamma}_2$, $\boldsymbol{\gamma}_3$), de manera que cada índice esté en una, y solo en una, de las tres sublistas. Por ejemplo



```python
gamma1 = (2,4,6,)
gamma2 = (1,7,)
gamma3 = (3,5,)
```

Si concatenamos las listas $\boldsymbol{\gamma}_1$, $\boldsymbol{\gamma}_2$ y $\boldsymbol{\gamma}_3$ (*las listas de Python se concatenan con la operación* `+`), tendremos una lista $\boldsymbol{\alpha}$ con todos los índices (aunque en un orden arbitrario):



```python
alpha = gamma1 + gamma2 + gamma3
alpha
```

Es decir
$$\boldsymbol{\alpha}=\boldsymbol{\gamma}_1\concat\boldsymbol{\gamma}_2\concat\boldsymbol{\gamma}_3$$
donde con $\concat$ indicamos la concatenación de las listas.



##### Creando la matriz identidad como suma de matrices



Consecuentemente, tenemos que la suma de las siguientes matrices
$$
\elemRPE{\Mat{I}}{\gamma1}\elemLPE{\Mat{I}}{\gamma1} + \elemRPE{\Mat{I}}{\gamma2}\elemLPE{\Mat{I}}{\gamma2} + \elemRPE{\Mat{I}}{\gamma3}\elemLPE{\Mat{I}}{\gamma3} $$
tiene que ser la matriz identidad. Veamoslo:



```python
(I(p)|gamma1)*(gamma1|I(p)) + (I(p)|gamma2)*(gamma2|I(p)) + (I(p)|gamma3)*(gamma3|I(p))
```

Por si no lo ve, visualicemos los distintos sumandos mediante sucesivos sistemas



```python
for lista in [gamma1, gamma2, gamma3]:
    display(Sistema([lista, I(p)|lista, lista|I(p), (I(p)|lista)*(lista|I(p))]) )
```

&#x2026; el primer elemento de cada sistema es la lista de índices $\boldsymbol{\gamma}_i$, el segundo es la correspondiente selección de las columnas de la matriz identidad $\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_i}$, el tercero la correspondiente selección de las filas de la matriz identidad $\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_i}$ y el cuarto la matriz producto $\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_i}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_i}$. 

Si suma las últimas matrices de cada sistema, verá que obtiene la matriz $\Mat{I}$ de orden 7.

**Recapitulando**

Con lo anterior tenemos que si $\boldsymbol{\alpha}$ es la concatenación de listas de índices $\boldsymbol{\gamma}_1$, $\boldsymbol{\gamma}_2,\ldots,\boldsymbol{\gamma}_k$ (por ejemplo, tres):
$$
\boldsymbol{\alpha}=\boldsymbol{\gamma}_1\concat\boldsymbol{\gamma}_2\concat\boldsymbol{\gamma}_3,
$$
de manera que en la lista completa $\boldsymbol{\alpha}$ están presentes todos los índices de $1$ a $p$, sin que ninguno esté repetido, entonces:
$$\Mat{I}= 
\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}=
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_1} + \elemRPE{\Mat{I}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_2} + \elemRPE{\Mat{I}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_3}$$

Vamos a usar estas igualdades, junto a las propiedades del producto de matrices, para encontrar otras formas de cálculo del producto de matrices.



##### Calculando la matriz producto $\Mat{AB}$ con las columnas de $\Mat{A}$ y las filas de $\Mat{B}$ reordenadas



Como
$$\Mat{AB}=\Mat{AIB}=
\Mat{A}\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}\Mat{B}=
%\Big(\Mat{A}\elemRPE{\Mat{I}}{\boldsymbol{\alpha}}\Big)\Big(\elemLPE{\Mat{I}}{\boldsymbol{\alpha}}\Mat{B}\Big)=
\elemRPE{\Mat{A}}{\boldsymbol{\alpha}}\elemLPE{\Mat{B}}{\boldsymbol{\alpha}}$$
tenemos que podemos reordenar arbitrariamente las columnas de $\Mat{A}$ y, con la misma reordenación, las filas de $\Mat{B}$, de manera que el producto de las dos nuevas matrices $\elemRPE{\Mat{A}}{\boldsymbol{\alpha}}\elemLPE{\Mat{B}}{\boldsymbol{\alpha}}$ es igual que el producto $\Mat{AB}$.

Generemos dos matrices $\Mat{A}$ de orden $m$, $p$ y $\Mat{B}$ de orden $p$, $n$;



```python
# por ejemplo 
m = 3
p = 7
A = Matrix( [ [ (1+i+j)  for i in range(p) ] for j in range(m) ] )

n = 2
B = Matrix( [ [ (1+j+i*j)  for i in range(n) ] for j in range(p) ] )
```

Reordenamos los `p` índices al azar:



```python
indices = random.sample(range(1, A.n+1), A.n)
print(indices)
```

Representamos las matrices originales



```python
Sistema([A,B])
```

Representamos las matrices reordenadas



```python
Sistema([A|indices, indices|B])
```

Y calculemos ambos productos: $\Mat{AB}$ a la izquierda y $\elemRPE{\Mat{A}}{\boldsymbol{\alpha}}\elemLPE{\Mat{B}}{\boldsymbol{\alpha}}$ a la derecha&#x2026; ¡son iguales!



```python
Sistema([A*B, (A|indices)*(indices|B),])
```

##### Calculando la matriz producto como suma de matrices



Como
$$\Mat{AB}=\Mat{AIB}=
\Mat{A}\Big(
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_1} + 
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_2} + 
\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_3}
\Big)\Mat{B}=
\Mat{A}\Big(\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_1}\Big)\Mat{B} + \Mat{A}\Big(\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_2}\Big)\Mat{B} + \Mat{A}\Big(\elemRPE{\Mat{I}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{I}}{\boldsymbol{\gamma}_3}\Big)\Mat{B}=
\elemRPE{\Mat{A}}{\boldsymbol{\gamma}_1}\elemLPE{\Mat{B}}{\boldsymbol{\gamma}_1} + 
\elemRPE{\Mat{A}}{\boldsymbol{\gamma}_2}\elemLPE{\Mat{B}}{\boldsymbol{\gamma}_2} + 
\elemRPE{\Mat{A}}{\boldsymbol{\gamma}_3}\elemLPE{\Mat{B}}{\boldsymbol{\gamma}_3} $$
tenemos que podemos calcular $\Mat{AB}$ mediante sumas de productos de submatrices.

Ahora considere dos matrices $\Mat{A}$ de orden $m$, $p$ y $\Mat{B}$ de orden $p$, $n$;



```python
m = 3
p = 6
A = Matrix( [ [ (1+i+j)  for i in range(p) ] for j in range(m) ] )
A
```

```python
n = 2
B = Matrix( [ [ (2*(j-2*i))  for i in range(n) ] for j in range(p) ] )
B
```

Ahora podemos considerar los índices pares por una parte, y los impares por otra



```python
pares   = [2,4,6]
impares = [1,3,5]
```

Representamos las submatrices seleccionadas con los índices pares, así como su producto



```python
Sistema([A|pares, pares|B, (A|pares)*(pares|B)])
```

Y hacemos lo mismo empleando los índices impares



```python
Sistema([A|impares, impares|B, (A|impares)*(impares|B)])
```

Y ahora calculemos $\Mat{AB}$ por una parte, y sumemos 
$
\elemRPE{\Mat{A}}{pares}\elemLPE{\Mat{B}}{pares} +\elemRPE{\Mat{A}}{impares}\elemLPE{\Mat{B}}{impares}
$



```python
Sistema([A*B, (A|pares)*(pares|B)+(A|impares)*(impares|B)])
```
