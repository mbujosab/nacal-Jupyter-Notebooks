## Suma de vectores



Genere los vectores $\Vect{a}$ y $\Vect{b}$:

-   El vector $\Vect{a}$ está formado por la lista de números: 1, 4, 9, 2.
-   El vector $\Vect{b}$ está formado por la lista de números: 1, 2, 1, 0.



```python
a = Vector([1, 4, 9, 2])
b = Vector([1, 2, 1, 0])
```

Sume ambos vectores:



```python
a + b
```

Después, obtenga la tercera componente de la suma de ambos: $\quad \elemR{(\Vect{a}+\Vect{b})}{3}$.



```python
(a + b)|3
```

Ahora sume la tercera componente de $\Vect{a}$ con la tercera de $\Vect{b}$: $\quad \eleVRPE{a}{3}+\eleVRPE{b}{3}$.   
Compruebe que obtiene el mismo resultado que obtuvo en la celda anterior. 
¿Cómo se llama esta propiedad? (*encontrará el nombre de la propiedad en el [libro](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.2)*)



```python
(a|3) + (b|3)
```

### Orden de ejecución de operaciones y necesidad de poner paréntesis



Ahora pruebe a quitar los paréntesis de la expresión anterior y ejecute el código
(*quite la almohadilla `#` y ejecute el siguiente código para verlo*).



```python
# a|3 + b|3
```

Python arrojará un error, puesto que lee y ejecuta las operaciones consecutivamente de izquierda a derecha: por tanto, al llegar al símbolo `+` trata de sumar el vector $\Vect{b}$ a la tercera componente de $\Vect{a}$ (el número `9` que ya ha calculado), pero la suma de números con vectores no está definida y se produce el error. Así pues, es necesario usar paréntesis para indicar el orden correcto de las operaciones. Esto permite calcular primero lo que hay dentro de los paréntesis y luego realizar la suma de los dos resultados.

Tenga en cuenta que cuando los humanos leemos $\;\eleVR{a}{3}+\eleVR{b}{3},\;$ interpretamos mentalmente que debe ser $\;\eleVRPE{a}{3}+\eleVRPE{b}{3};\;$, pero Python requiere que le indiquemos claramente el orden de operaciones con paréntesis.

Cuando las expresiones son complejas, incluso los humanos necesitamos paréntesis para interpretar correctamente. Siempre que escriba una expresión matemática y tenga dudas, es mejor usar paréntesis. Recuerde que, al trabajar con Python, las operaciones se ejecutan de izquierda a derecha, y por lo tanto, los paréntesis son normalmente necesarios.

Con el producto ocurre algo similar. Nosotros comprendemos la expresión: $$2\Vect{a}$$ porque realmente interpretamos: $$2\cdot\Vect{a}$$ (fíjese en el punto que indica el *producto*).
Python no funciona así. Si usted teclea `2a`, Python le mostrará un error. 
Compruébelo en la siguiente celda:



```python
# 2a
```

Python requiere explícitamente la indicación de operación producto con el símbolo `*`.



```python
2*a
```

*Observación*: Tal como se ha programado el módulo [NAcAL](https://github.com/mbujosab/nacallib), admite la multiplicación por ambos lados&#x2026; 
así se mantiene la asociatividad por la izquierda del operador selector (tal como se explica en el [libro](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.1.4.4)).



```python
a*2
```

Pero tenga en cuenta que una expresión como $\Vect{x}2$ nunca aparece en los libros de texto. 
El modo convencional de escribir dicho producto es $2\Vect{x}$, es decir, *siempre con el coeficiente delante*.



Obtenga la tercera componente del doble del vector $\Vect{a}$, es decir: $\elemRP{2\Vect{a}}{3}$



y ahora obtenga la tercera componente de $\Vect{a}$ y multiplíquela por 2: $\quad 2\elemRPE{\Vect{a}}{3}$.



Observe que en ambos casos se obtiene el mismo resultado.
