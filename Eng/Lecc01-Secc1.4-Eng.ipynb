## Matrices of $\R[m\times n]$



### Defining matrices



The notation indicated in the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.4) establishes that by enclosing a list of vectors (with the same number of components) between brackets, one denotes the matrix whose columns are these vectors. [NAcAL](https://github.com/mbujosab/nacallib) implements this convention by using a list or tuple of vectors (with the same number of components) as an argument for the class `Matrix()`.

Thus, by enclosing the vectors $(1, 2,)$, $(1, 0,)$, and $(9, 2,)$ in brackets, we denote the matrix of $\R[2\times 3]$:

$$\begin{bmatrix}1&1&9\\2&0&2\end{bmatrix}$$



With [NAcAL](https://github.com/mbujosab/nacallib), we achieve the same by placing this list of vectors as an argument of the class `Matrix( )`



```python
a = Vector( [1,2] )
b = Vector( [1,0] )
c = Vector( [9,2] )

D = Matrix( [a,b,c] )
```

```python
print(D)              # representation of the matrix in the console
display(D)            # representation of the matrix in Jupyter Notebook
```

The representation with `repr(D)` explicitly indicates that `D` is a matrix formed by a list of vectors, each formed by a list of numbers:



```python
repr(D)              # more explicit representation
```

#### An alternative way to instantiate different from that of the book, but more convenient when working on a computer



We can instantiate a matrix using a list (or tuple) of lists or tuples. In this case, we describe the matrix row by row:



```python
A = Matrix( ( [1,1], [2,4], [3,9] ) )
A
```

This alternative way of defining matrices (*row by row* instead of *column by column*) proves to be more convenient when using [NAcAL](https://github.com/mbujosab/nacallib) (although it is not the one described in the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.4)).



### Some attributes and procedures of `Matrix()`



#### Attribute: `.lista`



[NAcAL](https://github.com/mbujosab/nacallib) stores the ordered list of vectors (*of columns*) of each `Matrix()` in its attribute `.lista`.



```python
print( A.lista )
```

Note that, although we defined the matrix `A` *row by row*, the attribute `.lista` stores the corresponding list of columns (thus respecting the definition of matrix given in the course).



#### The Order of the matrix (attributes: `.m` and `.n`)



The attribute `.m` tells us the number of rows of the matrix, while the attribute `.n` indicates its number of columns. Thus, the order of $\Mat{A}$ is



```python
print( A.m )
print( A.n )
```

#### Method: `.sis()`



We can represent a matrix in the generic form of a `Sistema()` using `.sis()`.



```python
A.sis()
```

```python
print(A.sis())
```

Note that `repr(A.sis())` indicates that `A.sis()` is a system formed by a list of vectors.



```python
repr(A.sis())
```

And that `repr(A)` shows that `A` is a matrix (that is, a system) formed by the same list of vectors.



```python
repr(A)
```

Thus, although `A.sis()` and `A` are objects of different types



```python
B = A.sis()
print(type(A))
print(type(B))
```

and with different representation in `NAcAL`,



```python
# We create a system with both objects and text strings to see A and B together
Sistema( ['The matrix: ', A, 'The system: ',B] )
```

in reality, they are the same mathematical object (ordered lists of vectors). Therefore, if we ask Python whether `A` and `B` are equal, it responds that they are.



```python
B==A
```

## The transpose



(*[Section 1.4.5](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.1.4.5) of the book*)

Using the tools described above, we have a way to transpose $\Mat{A}$: 
by introducing as an argument of `Matrix()` the list of the lists corresponding to its columns (which is how to describe in [NAcAL](https://github.com/mbujosab/nacallib) a matrix row by row).

Thus, if the matrix is: 
$$\Mat{D}=\begin{bmatrix}1&1&9\\2&0&2\end{bmatrix}$$ then its transpose is:



```python
Matrix( [ (D|1).lista, (D|2).lista, (D|3).lista ] )
```

This can be automated with a loop that traverses the rows of the matrix (using [list comprehension](https://ellibrodepython.com/list-comprehension-python)):



```python
B = Matrix ( [ (D|j).lista for j in range(1,D.n+1) ] )
B
```

If you study the code of [NAcAL](https://github.com/mbujosab/nacallib), you will see that transposition is programmed as such.



I searched for a symbol in Python to indicate the operation of *transposition* in the style of the symbol $^\T$ used in the course. 

Unfortunately, Python has a strong limitation concerning the availability of symbols to be used as ["magic methods"](https://realpython.com/python-magic-methods/), so I found no better solution than to use the symbol (`~`) for transposition (instead of the symbol $^\T$). 

Unfortunately, this symbol must be placed *in front of* the matrix that one wants to transpose. 
Therefore, in [NAcAL](https://github.com/mbujosab/nacallib), the transpose of $\MatT{D}$ is written as: `~D`



```python
~D
```

Let us verify that upon transposing, the indices of the elements of the matrix change sides:



```python
1|(D)|3
```

```python
3|(~D)|1
```

```python
3|(~D) == D|3 # the third row of the transposed D equals the third column of D
```

```python
(~D)|2 == 2|D # the second column of the transposed D equals the second column of D
```

If we want to obtain a matrix whose only *column* is the second row of $\Mat{D}$, we just need to instantiate a `Matrix` using a list containing that row:



```python
Matrix( [2|D] )
```

## Definition of some special matrices



(*[Section 1.4.6](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.1.4.6) of the book*)



### Null matrix



With `M0(m,n)` we generate a null matrix of order $m$ by $n$. For example



```python
M0(2,3)
```

### Identity matrix



With `I(m)` we generate an identity matrix of order $m$. For example



```python
I(2)
```

### Opposite matrix



With the sign `-` in front of a matrix, we obtain its opposite



```python
-A
```

### Diagonal matrix



The method `.diag()` of `Vector()` returns a square diagonal matrix whose main diagonal contains the components of the employed vector.



```python
v = Vector( [1,2,3,4] )
v.diag()
```

In turn, the method `.diag()` of `Matrix()` returns a `Vector()` with the components of the main diagonal of the matrix.



```python
Matrix( ( [1,1,1],[2,4,8], [3,9,0] ) )
```

```python
Matrix( ( [1,1,1],[2,4,8], [3,9,0] ) ).diag()
```

```python
B = Matrix( ( [1,1,1], [2,4,8], [3,9,0] ) )
B
```

Note that in Python `(2)` is the number `2` in parentheses, but `(2,)` is a tuple with a single component (the `2`). 

The difference is the comma "`,`" after the number. 

In the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf), the same notation convention has been adopted to distinguish between numbers and vectors of $\R[1]$.



```python
type( (2) )
```

Here the right index is within a tuple, so we create a *matrix* column:



```python
(3,2,1,)|B|(3,)
```
