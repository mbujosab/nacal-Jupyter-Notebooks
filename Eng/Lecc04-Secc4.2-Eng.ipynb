## Pre-echeloning a matrix using the left-to-right elimination method



The following function `ppivote` returns the position of the first non-zero element of a vector (when the vector is the only argument).



```python
ppivote = lambda v, k=0: ( [i for i,c in enumerate(v.lista, 1) if (c!=0 and i>k)] + [0] )[0]
```

```python
ppivote(Vector([0,4,0,0,3,0]))
```

If the function is provided with a second argument representing a position from which to start searching, it returns the position of the first non-negative component starting from the position following the indicated one.



```python
ppivote(Vector([0,4,0,0,3,0]), 2)
```

When the function `ppivote` does not find a non-zero component, it returns the value zero.



```python
ppivote(Vector([0,4,0,0,0,0]), 2)
```

```python
x = Vector((1, 1, -2)); y = Vector((1, -2, 1)); z = Vector((-2, 1, 1))
A = Matrix([x,y,z])
A
```

Using the function `ppivote` and elementary transformations, we will echelon the matrix `A`.



```python
M = Matrix(A)            # Create a copy of A to avoid losing the original matrix
r = 0                    # Pivot counter
for i in range(1,M.m+1): # For each row: 
    p = ppivote((i|M),r) #   find the position of the first non-zero component to the right of column r 
    if p > 0:            #   if we have found one (if the row is not filled with zeros to the right)
        r += 1           #   the counter adds a new pivot to the count
        M & T( {r, p} )  #   we swap the r-th column with the column where we found the pivot 
        M & T( [(fracc(-(i|M|j),(i|M|r)), r, j) for j in range(1,M.n+1) if j > r] ) # with the pivot we nullify all
                                                                                    # components to its right

M  # finally, we display the result (note that M is no longer a copy of A)
```

For the matrix
$$\begin{bmatrix}1&1&-2\\1&-2&1\\-2&1&1\end{bmatrix}$$
the steps performed by the algorithm are:
$$\underset{\begin{subarray}{c} \left[\left(-1\right)\mathbf{1}+\mathbf{2} \right]\\\left[\left(2\right)\mathbf{1}+\mathbf{3} \right]\\\left[\left(1\right)\mathbf{2}+\mathbf{3} \right]\end{subarray}}{\mathbf{\tau}}$$
that is,
$$\begin{bmatrix}1&1&-2\\1&-2&1\\-2&1&1\end{bmatrix}\xrightarrow{\underset{\begin{subarray}{c} \left[\left(-1\right)\mathbf{1}+\mathbf{2} \right]\\\left[\left(2\right)\mathbf{1}+\mathbf{3} \right]\end{subarray}}{\mathbf{\tau}}}\begin{bmatrix}1&0&0\\1&-3&3\\-2&3&-3\end{bmatrix}\xrightarrow{\underset{\begin{subarray}{c} \left[\left(1\right)\mathbf{2}+\mathbf{3} \right]\end{subarray}}{\mathbf{\tau}}}\begin{bmatrix}1&0&0\\1&-3&0\\-2&3&0\end{bmatrix}$$

(where we have excluded from the scheme the swaps `T({r,p})` when `p==r`, since they do not modify the matrix).

Try modifying the preceding code to achieve new variants of it.

A similar code to the above is implemented in our library in the class `Elim` (*Elimination*)



```python
Elim(A)
```

&#x2026; and has some interesting additions. Let’s see it with the following matrix `B`.



```python
B = Matrix([[0,1,7,5],[4,3,2,1],[1,1,fracc(9,4),1]])
B
```

If we give it a non-empty second argument different from `0` or `False`, it will show us the steps taken during the elimination:



```python
L = Elim(B, True)
```

In this representation, both unnecessary swaps and unnecessary steps that eliminate zeros to the right of the pivot are excluded&#x2026; note that the algorithm traverses the row from the pivot <u>to the end</u>, operating even when the component `i,j` is a zero.  
But only the operations that actually modify the matrix are shown.

Note that `Elim` results in a pre-echelon form, but the original matrix does not change:



```python
B
```

### Examples of different elimination variants



**Rules followed in the previous example, but which are optional**

-   The method above, instead of subtracting $3/4$ of the first column from the second, multiplies the second column by $4$ and then subtracts $3$ times the first column from the second (new column). This way, it avoids introducing rational numbers into the matrix.

But if we do not mind generating rational components, we can take fewer steps by directly adding a rational amount to one vector from another. 

-   The method above searched for the first non-zero number in the first column (the pivot of the first column) and nullified the elements remaining to its right. It started with the first column of the matrix and followed the same pattern, moving left to right across all the columns of the matrix (which in the book has been referred to as [*left-to-right elimination*](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.4.2.1)). However, it could have started with the last column applying elimination from right to left. It could also have sought the last non-zero number in the last column and used it to nullify the values to its right.

-   The method above only nullified the elements of each pivot’s row that were to its right, but it could have also nullified those to its left.

-   The method above performs operations with the columns. However, most textbooks apply the elimination between rows from top to bottom (*we will see row operations later*). Or we could have eliminated the values to the right of each pivot from the right (using column operations), and the values below (using row operations).

-   The method above does not reorder the columns so that the pivots appear in a descending staircase arrangement with null columns to the right of non-null ones. But Gaussian elimination does.

-   The method above does not normalize the pivots to be one. However, Gauss-Jordan elimination does.

In other words, there are many ways to apply elementary transformations to a matrix to achieve another matrix with more zeros than the original matrix. `Elim(B)` implements a very specific way of doing this.

Internally, `Elim(B)` creates a copy of the matrix `B` and calls the procedure `elim()` to execute the operations in a very specific manner. 
But the procedure `elim()` has many ways to implement the elimination. 
The choice of each of these methods is made with the argument `variant`.

-   **`variant`:** allows selecting the variant of elimination we wish to apply:
    
    **Possible values**.
    
    -   0 (by columns, left to right): This is the default option and avoids adding multiples of one column to another if that multiple results from multiplication by a rational number.
    -   +1 (fast reduction): Performs fewer steps, as it nullifies each element with a single Type I operation (but as a trade-off, rational numbers may appear in the process).
    -   +2 (double reduction): If the elimination is performed with column operations, it makes zeros to the right *and to the left* of each pivot (if it is by rows, it makes zeros above and below each pivot).
    -   +4 (by rows): Row operations (if nothing else is indicated, the elimination goes from top to bottom using the first non-zero component of each row as a pivot and avoiding fractions if possible).
    -   +10 (normalization of the pivots): Once the matrix is pre-echeloned, Type II transformations are applied to normalize the pivots so that all have the value `1`.
    -   +20 (echelone the matrix): Once the matrix is pre-echeloned, the columns (or rows) are reordered to obtain a staggered form.
    -   +100 (backward): If operating by columns: right-to-left elimination (using the last non-zero element of each column as a pivot). If operating by rows: bottom-up elimination (using the last non-zero element of each row as a pivot).

By default `arg = 0`: (simple forward reduction, by columns, avoiding fractions).

We indicate whether we want the steps to be displayed with the argument `rep`:

-   **rep:** representation with $\LaTeX$ of the steps taken. Default value `rep=False` (does not show the steps). Any non-empty value other than `0` or `False` will cause the elimination steps to be displayed.

The way to manage the different variants is to sum the numbers for each option; this number corresponds to the type of elimination desired.

For example, `1+2+4+100=107` is elimination by fast reduction, on both sides of each pivot, by rows (therefore, double reduction means there are zeros above and below each pivot, and backward means top-to-bottom here). Let’s see it:



```python
B.elim(107, rep=True)
```

But `2+10=12` is elimination by double reduction by columns (that is, zeros to the right and left of each pivot), normalizing the pivots to be ones:



```python
B.elim(12, rep=True)
```

Hence, the Gauss-Jordan elimination, which also scales, would be `2+10+20=32`



```python
B.elim(32, rep=True)
```

Although the procedure `elim()` has great versatility, elimination is typically performed to obtain a specific type of matrix. Therefore, certain shortcuts are programmed. 

In cases of column elimination, it is common to use one of the following:



```python
display(B.K()) # performs steps to achieve a pre-echeloned form by columns
display(B.L()) # performs steps to achieve a echeloned form by columns
display(B.R()) # performs steps to achieve a reduced echeloned form by columns
```

In cases of row elimination, it is common to use one of the following:



```python
U = display(B.U()) # performs steps to achieve a echeloned form by rows
UR = display(B.UR()) # performs steps to achieve a reduced echeloned form by rows
```

Other ways to achieve the first three pre-echelon forms by columns are:



```python
K = Elim(B)   # pre-echeloned form by columns (left-to-right elimination). Equivalent to B.k()
L = ElimG(B)  # echeloned form by columns (Gaussian elimination). Equivalent to B.L()
R = ElimGJ(B) # reduced echeloned form by columns (Gauss-Jordan elimination). Equivalent to B.R()
Sistema([K,L,R])
```

And the equivalents by rows are:



```python
KF = ElimF(B)   # pre-echeloned form by rows (top-to-bottom elimination)
U = ElimGF(B)  # echeloned form by rows (Gaussian elimination). Equivalent to B.U()
RF =ElimGJF(B) # reduced echeloned form by rows (Gauss-Jordan elimination). Equivalent to B.UR()
Sistema([KF,U,RF])
```

## Row Transformations



*First, read section [4.A](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.alph3.4.Alph1) of the book.*

When elementary transformations act from the left side of the matrix, they perform operations with the rows.

For example, we can swap the first two rows `C` as follows:



```python
display(B)
T( {1,3} ) & B.copy()
```

(*Remember that if you apply elementary transformations directly on a matrix, it modifies the matrix. If you want to keep the original matrix unchanged, apply the transformations to a copy*).

Please note that when several transformations act from the left, the first to act are those closest to the matrix; thus, both



```python
T({1,2}) & T((2,2)) & Matrix(A)
```

and



```python
T([ {1,2}, (2,2) ]) & Matrix(A)
```

are equivalent to (and please note the use of parentheses):



```python
T({1,2}) & ( T((2,2)) & Matrix(A) )
```

that is, in all three cases, the second row is multiplied by 2 first (operation within the parentheses in the last example, but nearest to the matrix in the first two) and then the first two rows are swapped.



### Product of an elementary matrix on the left of another matrix.



Just as with transformations of the columns, we can transform the rows of a matrix by transforming the rows of the identity matrix and then multiplying the resulting matrix by the matrix we wish to transform:



```python
T((40,3,2)) & B.copy()
```

```python
elementaryMatrix =  T((40,3,2)) & I(3)
elementaryMatrix
```

```python
elementaryMatrix * B
```

Note that the notation does not work the same way when the elementary transformation is applied to the rows of the identity. In the example, the `40` is not in the third row, second column. If you notice, the elementary matrix we obtained by transforming the rows of $\Mat{I}$ is the transpose of the elementary matrix we would have obtained if we transformed the columns of the identity. That is, when applying the elementary transformation from the left, we transpose the elementary matrix we would obtain if we applied the transformation from the right to $\Mat{I}$.



### Row and Column Transformations



If we have transformations acting on the left and on the right, we operate on both rows and columns.



```python
T( {1,2} ) & T( (2,2) ) & Matrix(A) & T( [(-3,2,3), (10,1), {2,3}] )
```

In fact, the procedure `elim` allows for chaining different types of transformations. For example, the following code applies row elimination (top to bottom). Next, column elimination (left to right). Then it normalizes the pivots and finally echelon the matrix.



```python
B.elim(4).elim(0).elim(10).elim(20,rep=True)
```

## Deciding the operations on your own



Let’s consider the following matrix



```python
D = Matrix([[0,7,1],[-4,21,3]])
D
```

If we ask `NAcAL` to pre-echelon the matrix, it will do the following:



```python
ElimG(D, True)
```

&#x2026; which is not very elegant, as the library follows fixed rules that may lead to performing more operations than necessary.

In this example, since the pivot of the last column is a `1`, it is much simpler to:

1.  swap the first column with the last,
2.  subtract seven times that column from the second, and since it will be nullified, swap the second column with the third.

That is, to apply the following sequence of transformations on the columns:



```python
TrC = [ T( {1,3} ),
        T( [(-7,1,2), {2,3}] ) ]
```

To see this, we will use `dispElim(Matriz, pasos)`, where `pasos` is a list of two components. 

-   The first corresponds to the transformations of the rows (in this example it will be an empty list since there are no transformations affecting the rows).
-   The second corresponds to the transformations on the columns.

These components can be: 

-   a list made up of objects of the class `T`, in which case all the "packaged" transformations in each `T` are shown as a single step in the succession of matrix changes.
-   An object of the class `T`, in which case we will see the application of elementary transformations and swaps one by one.

Since `TrC` is a list of `T` objects, we will see the elimination in steps 1. and 2. described above:



```python
dispElim( D, [[], TrC]  )
```

If the steps component consists of a single object of the class `T`, we will see how the matrix changes with each operation, one at a time:



```python
columnOperations = T( [ {1,3}, (-7,1,2), {2,3} ] )

dispElim( D, [[], columnOperations]  )
```

## Transposition of a sequence of elementary transformations



We say that we transpose a sequence of transformations by reversing the order of the transformations in the sequence. 
Just as with matrices, we transpose with the operator `~` acting on the left of `T`.



```python
h=T([ (-2, 1, 2), (-3, 1, 3), (-4, 1, 4), (-5, 1, 5),  (-2, 2, 3), (4, 2, 4), (5, 2, 5) ])
h
```

```python
~h
```

Note that the order of the operations has been reversed after the transposition.

Consider the matrix obtained after applying a succession of transformations on the *columns of the identity matrix* of order $n$, resulting in a matrix. 
Well, the transpose of that succession applied to the *rows of the identity matrix* of order $n$ will yield the transpose of the previous matrix.



```python
I(5) & h
```

```python
(~h) & I(5)
```

Note that applying a succession of transformations is akin to <u>multiplying</u> by a matrix, so that if we apply it on the columns of $\Mat{B}$



```python
B
```

a succession of transformations



```python
Matrix(B) & h
```

the transpose of that resultant matrix



```python
~(Matrix(B)&h)
```

can be calculated with an operation reminiscent of the transpose of the product&#x2026; as it is the transpose of the transformation acting on the transpose of `B`



```python
~h & ~Matrix(B)
```
