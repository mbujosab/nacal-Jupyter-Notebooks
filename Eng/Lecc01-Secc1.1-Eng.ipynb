### Defining Vectors



In the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.1), it is stated that enclosing a list of $ n $ numbers in parentheses expresses a vector in $\mathbb{R}^n$;
for example, enclosing the list 20, 10, 15 in parentheses expresses a vector in $\mathbb{R}^3$

$$\begin{pmatrix}20,&10,&15,\end{pmatrix}$$

To define a vector in [NAcAL](https://github.com/mbujosab/nacallib), it is necessary to indicate the corresponding [list](https://docs.python.org/es/3/tutorial/introduction.html#lists) of numbers as an argument to the class `Vector()`. 
For example, with the list `[20, 10, 15]` we obtain



```python
Vector( [20, 10, 15] )
```

We will obtain the same result if, instead of using a [list](https://ellibrodepython.com/listas-en-python), we use the following [tuple](https://ellibrodepython.com/tuplas-python) `(20, 10, 15)`



```python
a = (20, 10, 15)
Vector( a )
```

The list of numbers corresponding to each `Vector` is stored in the attribute `.lista`



```python
v = Vector( [0,1,2,3,4,5] )
print(v.lista)
```

In fact, it is not necessary to use the function `print()`, you can simply write `v.lista` to view the list associated with the vector `v`



```python
v.lista
```

#### Representation of Vectors: the `.rpr` attribute of the `Vector` class



When the working environment is a Jupyter Notebook, [NAcAL](https://github.com/mbujosab/nacallib) displays vectors in column format by default. However, we can instruct Python that we want a horizontal representation of the `Vector`; it is sufficient to assign the value `'h'` to the attribute `rpr`. *Please note that this only affects the representation* (the vector does not change, as the associated list of numbers remains the same).

Let's make a copy of `a` with horizontal representation (`rpr='h'`).



```python
x = Vector(a, rpr='h')
print(x.rpr)
```

Let's plot the vector:



```python
x
```

To change the representation of a previously defined vector, simply modify the value of its attribute `rpr`.



```python
x.rpr = 'col'  # any value other than the string 'h' will represent in column
print(x.rpr)
```

Since the `rpr` attribute is no longer `'h'`, the vector `x` is now represented vertically:



```python
x
```

If we want to make a copy that replicates all attributes (including the representation), we will use the method `.fullcopy()`



```python
a = Vector( [1,2,3,4], 'h') # a has horizontal representation
b = a.fullcopy()
b
```

```python
Vector(a)
```

```python
c = a.copy()
c
```

#### Method `.sis()`



In the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.1), a vector is defined as a system (i.e., an ordered list) of numbers, and the corresponding notation for systems in general is described.

We can retrieve the representation of a `Vector` as a generic `System` using the method `.sis()`. The method `.sis()` returns the vector in the form of a generic `System` and, therefore, uses the corresponding generic representation (that is, enclosed in brackets with the elements followed by semicolons, as indicated in the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#chapter.1)).



```python
x.sis()
```

The console representation explicitly indicates that `x.sis()` is a system.



```python
print(x.sis())
```

We also have the representation with `repr` that even more explicitly indicates that `x.sis()` is a system.



```python
repr(x.sis())
```

The function `type()` returns the type of the object.



```python
type(x.sis())
```

### Defining a Vector from the Components of Another



*Refer to Section 1.A of the appendix of Lesson 1 in the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.alph1.1.Alph1).*

By selecting several components from a `Vector` (with a tuple), we will obtain a new `Vector`. Starting from



```python
a = Vector( [10, 20, 30, 40, 50], 'h' )
display(a)
```

we will define a *sub-vector* by simultaneously selecting several components:



```python
a|(1,3,5)
```

We can repeat the same component multiple times.



```python
a|(5,1,1,1,1,1,5)
```

Note that by using a tuple of indices, we can easily reorder the components of a vector



```python
a|(5,4,3,2,1)
```

## Equality of Vectors



With `v = Vector( a )`, we create a vector with vertical representation whose list of components is that of `a`.
With `w = Vector( a, rpr='h' )`, we generate another vector with the same list, but with representation in row form. 

*Two vectors are equal if their corresponding ordered lists of numbers are equal* 
(regardless of whether the representation is vertical or horizontal).

The operator `==` checks whether two objects are equal (`True`) or not (`False`):



```python
v = Vector( a )
w = Vector( a, rpr='h' )
v == w
```

```python
b = Vector( [-2, 0, -2] )
a == -b
```

Two zero vectors with a different number of components are not equal



```python
V0(2) == V0(3)
```

With `!=` we check whether it is true that two vectors are distinct



```python
a != b      # Are they distinct?
```
