## Matrix Equality



Two matrices are equal if their corresponding systems are equal.



```python
a = Vector( [1,2] )
b = Vector( [1,0] )
c = Vector( [9,2] )
```

```python
A = Matrix( [a,b,c] )
A
```

```python
B = Matrix( [ [1,1,9], [2,0,2] ] )
B
```

```python
print(A|1 == B|1)
print(A|2 == B|2)
print(A|3 == B|3)
```

The same check as above but using [list comprehension](https://ellibrodepython.com/list-comprehension-python):



```python
[A|i == B|i for i in range(1,A.n+1)]
```

Consequently, `A` and `B` are equal.



```python
A == B
```

(*If the argument used with the selector is an index, we obtain a vector, but when selecting a tuple of indices we obtain a matrix*.)



```python
(1,)|A
```

The element in the first row and third column is:



```python
1|A|3
```

But if we transpose the matrix, we need to switch the indices of that component:



```python
3|(~A)|1
```

```python
~~A
```
