### Definiendo vectores



En el [libro](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.1) se indica que encerrando entre paréntesis una lista de $ n $ números expresamos un vector de $\mathbb{R}^n$;
por ejemplo, encerrando la lista 20, 10, 15 entre paréntesis expresamos un vector de $\mathbb{R}^3$

$$\begin{pmatrix}20,&10,&15,\end{pmatrix}$$

Para definir un vector en [NAcAL](https://github.com/mbujosab/nacallib) es necesario indicar la correspondiente [lista](https://docs.python.org/es/3/tutorial/introduction.html#lists) de números como argumento de la clase `Vector()`. 
Por ejemplo, con la lista `[20, 10, 15]` obtenemos



```python
Vector( [20, 10, 15] )
```

Obtendremos idéntico resultado si, en lugar de usar una [lista](https://ellibrodepython.com/listas-en-python), usamos la siguiente [tupla](https://ellibrodepython.com/tuplas-python) `(20, 10, 15)`



```python
a = (20, 10, 15)
Vector( a )
```

La lista de números correspondiente a cada `Vector` está almacenada en el atributo `.lista`



```python
v = Vector( [0,1,2,3,4,5] )
print(v.lista)
```

Realmente, no es necesario usar la función `print()`, basta con escribir `v.lista` para ver la lista asociada al vector `v`



```python
v.lista
```

#### Representación de Vectores: el atributo `.rpr` de la clase `Vector`



Cuando el entorno de trabajo es un Notebook de Jupyter, [NAcAL](https://github.com/mbujosab/nacallib) muestra los vectores en forma de columna por defecto. Pero podemos indicar a Python que queremos una representación horizontal del `Vector`; basta asignar el valor `'h'` al atributo `rpr`. *Tenga en cuenta que esto solo afecta a la representación* (el vector no cambia, ya que la lista de números asociada sigue siendo la misma).

Hagamos una copia de `a` con representación horizontal (`rpr='h'`).



```python
x = Vector(a, rpr='h')
print(x.rpr)
```

Pintemos el vector:



```python
x
```

Para cambiar la representación de un vector previamente definido, basta cambiar el valor de su atributo `rpr`.



```python
x.rpr = 'col'  # cualquier valor distinto de la cadena 'h' representará en columna
print(x.rpr)
```

Como el atributo `rpr` ya no es `'h'`, ahora el vector `x` es representado en vertical:



```python
x
```

#### Método `.sis()`



En el [libro](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.1) se define un vector como un sistema (i.e., una lista ordenada) de números y se describe la correspondiente notación para los sistemas en general.

Podemos recuperar la representación de un `Vector` como `Sistema` genérico con el método `.sis()`. El método `.sis()` devuelve el vector en forma de `Sistema` genérico y, por tanto, usa la correspondiente representación genérica (es decir, entre corchetes y con los elementos seguidos por puntos y comas, tal y como se indica en el [libro](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#chapter.1)).



```python
x.sis()
```

La representación en modo consola indica explícitamente que `x.sis()` es un sistema.



```python
print(x.sis())
```

También tenemos la representación con `repr` que indica aún más explícitamente que `x.sis()` es un sistema.



```python
repr(x.sis())
```

La función `type()` nos devuelve el tipo de objeto.



```python
type(x.sis())
```

### Definiendo un vector a partir de las componentes de otro

## Igualdad de vectores



Con `v = Vector( a )` generamos un vector con representación vertical y cuya lista de componentes es la de `a`.
Con `w = Vector( a, rpr='h' )` generamos otro vector con la misma lista, pero con representación en forma de fila. 

*Dos vectores son iguales si lo son las correspondientes listas ordenadas de números* 
(independientemente de si la representación es vertical u horizontal).

El operador `==` comprueba si dos objetos son iguales (`True`) o si no lo son (`False`):



```python
v = Vector( a )
w = Vector( a, rpr='h' )
v == w
```

```python
b = Vector( [-2, 0, -2] )
a == -b
```

Dos vectores nulos con diferente número de componentes no son iguales



```python
V0(2) == V0(3)
```

Con `!=` comprobamos si es verdadero que dos vectores son distintos



```python
a != b      # ¿son distintos?
```
