## Dot Product



With the following code we generate two vectors in $\R[4]$.



```python
a = Vector([10,2,5,4])
b = Vector([-1,5,-3,4])
```

Now we create a list of the *component-wise products* of both vectors (iterating over the components from the first to the last). Then, using `sum`, we sum the elements of the list:



```python
cross_product_list = [(a|i)*(b|i) for i in range(1, 5) ]

print(cross_product_list)

sum(cross_product_list)
```

These are the operations performed by the dot product between two vectors in $\R[n]$



```python
a*b
```

## Linear Combination of Vectors



Consider the vectors
$$\Vect{x}=(1, 1, -2,),\quad \Vect{y}=(1, -2, 1,)\quad \text{and}\quad \Vect{z}=(-2, 1, 1,).$$
Verify that $2\Vect{x}+2\Vect{y}+2\Vect{z}=\Vect{0}$



```python
x=Vector([1,1,-2])
y=Vector([1,-2,1])
z=Vector([-2,1,1])

2*x + 2*y + 2*z
```

Write the vector $\Vect{x}=(1, 1, -2,)$ explicitly as a linear combination of the columns of the identity matrix of order 3 (remember that  $5\VectCPE{I}{1}$ is a product, so its transcription is `5*(I(3)|5)`, meaning that it is necessary to indicate the product operation with `*`).



### Product of a Matrix by a Vector



Consider the following matrix $\Mat{A}$ (created using [list comprehension](https://ellibrodepython.com/list-comprehension-python)).



```python
A = Matrix( [[i-j  for j in range(3)] for i in range(4) ]  ) # range iterates over the integers from 0 to p-1 (as seen in the first column)
A
```

Now calculate the following linear combination of the columns of $\Mat{A}$:
$$
2 \elemR{\Mat{A}}{1} - 3 \elemR{\Mat{A}}{2} - 2 \elemR{\Mat{A}}{3}
$$
explicitly using the columns of the matrix in the expression (that is, transcribing literally from the expression above to code `nacal`; remember that $2\elemR{\Mat{A}}{1}$ is a product, so its transcription is `2*(A|1)`, meaning that it is necessary to indicate the product operation with `*`).



The following vector contains the coefficients (the scalars that multiply the columns) of the linear combination.



```python
b = Vector( [2,-3,-2] )
```

We can calculate one by one the elements of this linear combination of columns. To do this, we calculate the dot product of each of the rows of $\Mat{A}$ by $\Vect{b}$. For example, the first number of the linear combination is:



```python
(1|A)*b
```

And the second:



```python
(2|A)*b
```

We can calculate all the elements iterating through all the rows of $\Mat{A}$ using [list comprehension](https://ellibrodepython.com/list-comprehension-python).



```python
Vector( [ (i|A)*b  for i in  range(1, A.m+1) ] )
```

However, notably, the most convenient way is to write the product of $\Mat{A}$ by the vector $\Vect{b}$



```python
A*b
```

The first method we used (the explicit linear combination) is literally the definition of matrix by vector that we have seen in the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.2.2.2) (and is how the library calculates it `A*b`).

The proof that `(i|A)*b` is equal to `i|(A*b)` is proposed as an exercise in the same section of the book.
