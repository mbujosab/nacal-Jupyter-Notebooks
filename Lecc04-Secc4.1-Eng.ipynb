## Elementary Transformations



We will consider two types of elementary transformations, denoted with tuples. 
To indicate that a particular tuple is an abbreviation for an elementary transformation, we will write it as an argument of the class `T`.

-   **Type I:** The elementary transformations of the first type add $a$ times the $i$-th vector to the $j$-th vector:
    $$
           \underset{\left[\left(a\right)\mathbf{i}+\mathbf{j} \right]}{\mathbf{\tau}}
           $$
    We will denote these transformations with a tuple of three components: `T( (a, i, j) )`.
    
    For example, the abbreviation of the elementary transformation that adds 3 times the second vector to the first vector is written as



```python
T( (3, 2, 1) )
```

Note that the index of the modified vector appears last; that this vector is being added a multiple of another vector, whose index appears second; and that the coefficient by which it is multiplied appears first.

-   **Type II:** The elementary transformations of the second type multiply the $j$-th vector by a number $a$.
    $$
           \underset{\left[\left(a\right)\mathbf{j}\right]}{\mathbf{\tau}}
           $$
    We will denote these transformations with a tuple of two components: `T( (a, j) )`.
    
    For example, the elementary transformation that multiplies the second vector by 5 is written as



```python
T( (5, 2) )
```

Once again, the index of the modified vector appears last, and the coefficient by which it is multiplied appears first.



## Elementary Transformations of the Rows or Columns of a Matrix



### Column Transformations



When we apply an elementary transformation from the right to a matrix, we modify its columns. The operation of an elementary transformation is indicated with the symbol `&`

For example, consider the matrix $\Mat{A}$



```python
A=Matrix([Vector((1,2)), Vector((0,3)), Vector((-2,2))])
A
```

We are going to add twice (2) the first column (1) of $\Mat{A}$ to its third column (3).



```python
A & T( (2,1,3) )
```

When we apply an elementary transformation to a matrix, the matrix changes. Now the matrix $\Mat{A}$ has been modified:



```python
A
```

If we want to apply transformations to a matrix without losing the original matrix, we need to make a copy. For example, let us define the following matrix $\Mat{A}$



```python
x = Vector((1, 1, -2)); y = Vector((1, -2, 1)); z = Vector((-2, 1, 1))
A = Matrix([x,y,z])
A
```

and let’s make a copy (with `Matrix(A)`) to which we will apply a couple of transformations



```python
B = Matrix(A)
B & T( (2,1,3) ) & T( (5, 2) )
```

Now $\Mat{B}$ has changed, but $\Mat{A}$ remains unchanged.



```python
A
```

### Sequence of Elementary Transformations



*Read the [section 4.1.5](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.4.1.5) of the book.*

It is possible to generate a sequence of elementary transformations:
$$
\underset{\begin{subarray}{c} \left[\left(a\right)\mathbf {j}\right]\\\left[\left(b\right)\mathbf{p}+\mathbf{q} \right]\end
{subarray}}{\mathbf{\tau}}
$$
To do this, we can use a list of abbreviations as an argument of the class `T`. 
For example:



```python
T( [(3, 2, 1), (5, 2)] )
```

#### Composition of Elementary Transformations



Furthermore, if we apply an elementary transformation to another elementary transformation, we achieve a transformation that is the composition of both. For example



```python
z = T( (2,1,3) ) & T( (5, 2) )
z
```

If we now apply `z` to a copy of `A`, we will obtain the same result as above



```python
C = Matrix(A)
C & z
```

Objects of the class `T` have an attribute `.abreviaturas`, where the list of abbreviations is stored. For example, in the attribute `.abreviaturas` of `z`, the list of abbreviations of the transformations applied above appears:



```python
z.abreviaturas
```

In this example, the attribute `.abreviaturas` consists of a list of tuples (with two or three elements), but later we will see that the list `.abreviaturas` can also contain sets of two indices `{i,j}` (to denote exchanges).



##### Instantiating the class `T` with a List of Abbreviations (or a List of Lists of Abbreviations or Objects of the Class `T`)



We have seen that it is possible to instantiate `T` using a single abbreviation as an argument, but also with a list of abbreviations. 
For example:



```python
T( [(-3,1,2), (10,1), (5,6,3), (-9,6)] )
```

By default, the list of operations appears in vertical format, but we can indicate that we want to see the list in horizontal format:



```python
T( [(-3,1,2), (10,1), (5,6,3), (-9,6)], rpr='h' )
```

We can also use as an argument:

-   a list of lists of abbreviations
-   another object of the class `T`
-   a list of objects of the class `T`
-   a list of lists of abbreviations and lists of `T` objects

Example using a list of lists



```python
Tr = T( [ [(-3,1,2), (10,1)], [(5,6,3), (-9,6)]], rpr='h' )
Tr
```

Example using an object of the class `T`



```python
T( Tr, rpr='h' )
```

Example using a list of objects of the class `T`



```python
TrC = [ T( [(-3,1,2), (10,1)] ),  T( [(5,6,3), (-9,6)] ) ]
T( TrC, rpr='h' )
```

Example using a list of abbreviations, lists of abbreviations, and lists of `T` objects



```python
T( [ (-3,1,2),  [T( [(10,1)] )],  [(5,6,3), (-9,6)] ], rpr='h'  )
```

The result is the same.



#### Exchanges



*Read the [section 4.1.6](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.4.1.6) of the book.*

There are multiple combinations of elementary transformations that can achieve the exchange of two vectors. The following is an example of a sequence that exchanges the first and third vectors:



```python
display(A)
intercambio = T( [ (-1,3), (-1,3,1), (1,1,3), (-1,3,1) ] )
Matrix(A) & intercambio
```

We can see the steps with the help of `dispElim()` (later we will see how `dispElim()` works):



```python
dispElim( A, [[], intercambio] )
```

Fortunately, there is a much simpler way to perform an exchange.
With a set of two indices, we indicate the exchange of the vectors corresponding to those indices. 
For example, `T( {s, t} )` exchanges the $s$-th and $t$-th vectors:
$$
\underset{\left[\mathbf{s}\rightleftharpoons\mathbf{t}\right]}{\mathbf{\tau}}
$$
Thus, to exchange the first and third columns of a copy of $\Mat{A}$, simply write



```python
Matrix(A) & T( {1,3} )
```

If the set contains only one index, the exchange does nothing:



```python
Matrix(A) & T( {1,1} )
```

## Elementary Matrices, Exchange Matrices, and Permutation Matrices



The application of a *single* elementary transformation to the identity matrix results in an elementary matrix.

-   **Type I Elementary Matrices:** Let’s see how applying the transformation `T(-5,1,3)` to the *columns* of the identity matrix of order 4 gives us a matrix resembling the identity, except that a $-5$ appears in position $(1,3)$.



```python
I(4) & T( (-5,1,3) )
```

But the same elementary transformation applied to the *rows* of the identity matrix gives us another distinct elementary matrix: the transposed version of the above elementary matrix:



```python
T( (-5,1,3) ) & I(4)
```

Therefore, the elementary transformations of *Type I* are associated with distinct elementary matrices depending on whether they act on columns (from the right) or on rows (from the left). Notice that one elementary matrix is the transpose of the other.

-   **Type II Elementary Matrices:** Now let’s see how applying the transformation `T(100,2)` to the *columns* of the identity matrix of order 3 gives us a matrix resembling the identity, except that a $100$ appears in the second position on the diagonal, that is, in position $(2,2)$.



```python
I(3) & T( (100,2) )
```

The *Type II* elementary matrices are symmetric; thus, we obtain the same elementary matrix when we act from the left on the identity matrix.



```python
T( (100,2) ) & I(3)
```

Note that the application of <u>a sequence of elementary transformations *does not* usually result in an elementary matrix</u>.



```python
I(3) & T( (2,1,3) ) & T( (5, 2) ) & T( (-5,3,1) ) & T ( (2,1,2) )
```

-   **Exchange Matrices:** The application of <u>a single exchange</u> between the columns (or between the rows) of the identity matrix results in an exchange matrix.



```python
I(4) & T( {2,4} )
```

```python
T( {2,4} ) & I(4)
```

-   **Permutation Matrices:** The application of a <u>sequence of exchanges</u> between the columns (and/or between the rows) of the identity matrix results in a rearrangement of the columns of the identity matrix that we call a permutation matrix.



```python
I(4) & T({3,1}) & T({2,3}) & T({1,2}) & T({3,4})
```

which is like an identity matrix, but with its rows or columns reordered.

Exchanges can be performed both from the left (rows) and from the right (columns).



```python
T( [ {1,2}, {3,4} ] ) & I(4) & T({3,1}) & T({2,3})
```

### Product of a Matrix by an Elementary Matrix



Verify with examples that applying an elementary transformation to the columns of a matrix is equivalent to multiplying that matrix by the corresponding elementary matrix; that is, that 
$$
{\Mat{A}}_{\mathbf{\tau}}=\Mat{A}\big({\Mat{I}}_{\mathbf{\tau}}\big)
$$



```python
# First we apply the transformation to the columns
Matrix(A) & T( (2,1,3) )
```

```python
# And now we multiply from the right by the corresponding elementary matrix
A * ( I(A.n) &  T((2,1,3)) )
```

Verify that this also occurs when acting from the left to transform the rows. First, we apply the transformation to the rows.



```python
# First we apply the transformation to the rows
T( (2,1,3) ) & Matrix(A)
```

```python
# And now we multiply from the left by the corresponding elementary matrix
( T((2,1,3)) & I(A.m) ) * A
```

### Transformations are Associative with Respect to Matrix Multiplication



Verify with examples that in general
$$
{\big(\Mat{A}\Mat{B}\big)}_{\mathbf{\tau}}=\Mat{A}\big({\Mat{B}}_{\mathbf{\tau}}\big)
$$
Remember to use a copy of the original matrices if you do not want them to change.



```python
# First the left side of the equality
( A * Matrix(B) ) & T((-5,2,1))
# Remember that if we do not want to modify the original matrices, we should act on copies of them.
```

```python
# Now the right side of the equality
A * ( Matrix(B) & T((-5,2,1)) )
# Remember that if we do not want to modify the original matrices, we should act on copies of them.
```

This holds true even when applying a sequence of elementary transformations (both from the right and from the left).

Let’s see an example of transformation of the columns:



```python
# We define a sequence of elementary transformations
st = T([(2, 1, 3), (5, 2), (-5, 3, 1), (2, 1, 2), {2,3}])
st
```

```python
( A * Matrix(B) ) & st
```

```python
A * ( Matrix(B) & st )
```

And now an example of transformation of the rows&#x2026;



```python
st & ( Matrix(A) *B )
```

```python
( st & Matrix(A) ) * B
```

## Elementary Transformations are Linear Operators



### Associative Property



Verify with examples that in general
$$
{\big(\lambda\Mat{B}\big)}_{\mathbf{\tau}}=\lambda\big({\Mat{B}}_{\mathbf{\tau}}\big)
$$
Remember to use a copy of the original matrices so that they do not change.



```python
# First the left side of the equality
( 5 * Matrix(A) ) & st
```

```python
# And then the right side
5 * ( Matrix(A) & st )
```

### Distributive Property



Verify with examples that in general
$$
{\big(\Mat{A}+\Mat{B}\big)}_{\mathbf{\tau}}={\Mat{A}}_{\mathbf{\tau}} + {\Mat{B}}_{\mathbf{\tau}}
$$



```python
# First the left side of the equality
( Matrix(A) + Matrix(B) ) & st
```

```python
# And then the right side of the equality
( Matrix(A) & st ) + ( Matrix(B) & st )
```

This is also true with transformations of the rows



```python
st & ( Matrix(A) + Matrix(B) )
```

```python
( st & Matrix(A) ) + ( st & Matrix(B) )
```
