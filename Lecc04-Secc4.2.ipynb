## Pre-escalonando una matriz por el método de eliminación de izquierda a derecha



La siguiente función `ppivote` de nuestra librería devuelve la posición del primer elemento no nulo de un vector (cuando el vector es el único argumento).



```python
ppivote = lambda v, k=0: ( [i for i,c in enumerate(v.lista, 1) if (c!=0 and i>k)] + [0] )[0]
```

```python
ppivote(Vector([0,4,0,0,3,0]))
```

Si a la función se le proporciona como segundo argumento una posición desde la que empezar a buscar, entonces devuelve la posición del primer componente no negativo a partir de la posición siguiente a la indicada.



```python
ppivote(Vector([0,4,0,0,3,0]), 2)
```

Cuando la función `ppivote` no encuentra un componente no nulo, devuelve el valor cero.



```python
ppivote(Vector([0,4,0,0,0,0]), 2)
```

```python
x = Vector((1, 1, -2)); y = Vector((1, -2, 1)); z = Vector((-2, 1, 1))
A = Matrix([x,y,z])
A
```

Usando la función `ppivote` y las transformaciones elementales, vamos a escalonar la matriz `A`.



```python
M = Matrix(A)            # Hacemos una copia de A para no perder la matriz original
r = 0                    # Contador de pivotes
for i in range(1,M.m+1): # En cada una de las filas: 
    p = ppivote((i|M),r) #   buscamos la posición del primer componente no nulo a la derecha de la columna r 
    if p > 0:            #   si hemos encontrado alguno (si la fila no está llena de ceros por la derecha)
        r += 1           #   el contador añade un nuevo pivote a la cuenta
        M & T( {r, p} )  #   intercambiamos la columna $r$-ésima con la columna donde encontramos el pivote 
        M & T( [(fracc(-(i|M|j),(i|M|r)), r, j) for j in range(1,M.n+1) if j > r] ) # con el pivote anulamos todos
                                                                                    # los componentes a su derecha

M  # por último mostramos el resultado (fíjese que M ya no es una copia de A)
```

Para la matriz
$$\begin{bmatrix}1&1&-2\\1&-2&1\\-2&1&1\end{bmatrix}$$
los pasos dados por el algoritmo han sido:
$$\underset{\begin{subarray}{c} \left[\left(-1\right)\mathbf{1}+\mathbf{2} \right]\\\left[\left(2\right)\mathbf{1}+\mathbf{3} \right]\\\left[\left(1\right)\mathbf{2}+\mathbf{3} \right]\end{subarray}}{\mathbf{\tau}}$$
es decir,
$$\begin{bmatrix}1&1&-2\\1&-2&1\\-2&1&1\end{bmatrix}\xrightarrow{\underset{\begin{subarray}{c} \left[\left(-1\right)\mathbf{1}+\mathbf{2} \right]\\\left[\left(2\right)\mathbf{1}+\mathbf{3} \right]\end{subarray}}{\mathbf{\tau}}}\begin{bmatrix}1&0&0\\1&-3&3\\-2&3&-3\end{bmatrix}\xrightarrow{\underset{\begin{subarray}{c} \left[\left(1\right)\mathbf{2}+\mathbf{3} \right]\end{subarray}}{\mathbf{\tau}}}\begin{bmatrix}1&0&0\\1&-3&0\\-2&3&0\end{bmatrix}$$

(donde hemos excluido del esquema los intercambios `T({r,p})` cuando `p==r`, pues no modifican la matriz).

Pruebe a realizar modificaciones del anterior código para lograr nuevas variantes del mismo.

Un código parecido al anterior está implementado en nuestra librería en la clase `Elim`  (*Eliminación Gaussiana*)



```python
Elim(A)
```

&#x2026; y posee algunos añadidos interesantes. Veámoslo con la siguiente matriz `B`.



```python
B = Matrix([[0,1,7,5],[4,3,2,1],[1,1,fracc(9,4),1]])
B
```

Si le damos un segundo argumento no vacío y distinto de `0` o `False`, nos mostrará los pasos dados en la eliminación:



```python
L = Elim(B, True)
```

En dicha representación se excluyen tanto los intercambios innecesarios como los pasos innecesarios que eliminan ceros a la derecha del pivote&#x2026; fíjese que el algoritmo recorre la fila desde el pivote <u>hasta el final</u>, operando incluso cuando el componente `i,j` es un cero.  
Pero solo se muestran las operaciones que realmente modifican la matriz.

### Ejemplos de las distintas variantes de eliminación



**Reglas que se han seguido en el ejemplo anterior, pero que son opcionales**

-   El método de más arriba, en lugar de restar $3/4$ de la primera columna a la segunda, multiplica la segunda columna por $4$ y luego resta $3$ veces la primera columna de la segunda (nueva columna). Gracias a ello, evita que aparezcan números racionales en la matriz.

Pero si no nos importa generar componentes racionales, podemos dar menos pasos añadiendo directamente una cantidad racional un vector a otro. 

-   El método de arriba ha buscado el primer número no nulo de la primera columna (el pivote de la primera columna) y ha anulado los elementos que quedaban a su derecha. Ha comenzado por la primera columna de la matriz y ha seguido la misma pauta avanzando de izquierda a derecha recorriendo todas las columnas de la matriz (lo que en el libro se ha llamado [*eliminación de izquierda a derecha*](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.4.2.1)). Pero podría haber comenzado por la última columna aplicando eliminación de derecha a izquierda. También podría haber buscado el último número no nulo de la última columna y haberlo usado para anular los valores de su fila a la derecha.

-   El método de arriba solo ha anulado los elementos de la fila de cada pivote que están a su derecha, pero podría haber anulado también los que están a su izquierda.

-   El método de más arriba realiza operaciones con las columnas. Pero la mayoría de manuales aplica la eliminación entre filas de arriba a abajo (*veremos las operaciones por columnas más adelante*). O incluso podríamos haber eliminado los valores a la derecha de cada pivote por la derecha (con operaciones por columnas) y los valores por debajo (con operaciones por filas).

-   El método de más arriba no reordena las columnas para que los pivotes aparezcan en una disposición de escalera descendente y que las columnas nulas queden a la derecha de las no nulas. Pero la eliminación gaussiana lo hace.

-   El método de más arriba no normaliza los pivotes para que sean unos. Pero la eliminación Gauss-Jordan lo hace.

Es decir, hay muchas formas de aplicar transformaciones elementales a una matriz para lograr otra matriz con más ceros que la matriz original. `Elim(B)` implementa una forma muy particular de hacerlo.

Internamente `Elim(B)` genera una copia de la matriz `B` y llama al procedimiento `elim()` para que ejecute las operaciones de una forma muy particular. 
Pero el procedimiento `elim()` tiene implementadas multitud de maneras de aplicar la eliminación. 
La elección de cada una de estas maneras se realiza con el argumento `variante`.

-   **`variante`:** permite escoger la variante de eliminación que queremos aplicar:
    
    **Posibles valores**.
    
    -   0 (por columnas, de izquierda a derecha) : Es la opción por defecto y evita añadir múltiplos de una columna a otra si dicho múltiplo es fruto de multiplicar por un número racional.
    -   +1 (reducción rápida) : Realiza menos pasos, pues anula cada elemento con una sola operación de Tipo I (pero a cambio, pueden aparecer números racionales en el proceso).
    -   +2 (doble reducción)  : Si la eliminación se realiza con operaciones sobre las columnas, hace ceros a la derecha *y a la izquierda* de cada pivote (si es por filas, hace ceros por encima y por debajo de cada pivote).
    -   +4 (por filas)        : Operaciones por filas (si no se indica nada más, la eliminación va de arriba a abajo usando como pivote el primer componente no nulo de cada fila y evitando fracciones si es posible).
    -   +10 (normalización de los pivotes) : una vez pre-escalonada la matriz, se aplican transformaciones de Tipo II para normalizar los pivotes de manera que todos tengan el valor `1`.
    -   +20 (escalona la matriz) : una vez pre-escalonada la matriz, reordena las columnas (o las filas) para obtener una forma escalonada.
    -   +100 (de atrás hacia delante) : Si opera por columnas: eliminación de derecha a izquierda (usando como pivote el último elemento no nulo de cada columna). Si opera por filas: eliminación de abajo a arriba (usando como pivote el último elemento no nulo de cada fila).

Por defecto `arg = 0`: (reducción simple hacia delante, por columnas y evitando fracciones).

Indicamos si queremos que se muestren los pasos con el argumento `rep`:

-   **rep:** representación con $\LaTeX$ de los pasos dados. Valor por defecto `rep=False` (no muestra los pasos). Cualquier valor distinto de `False` ocasiona que se muestren los pasos de eliminación.

La forma de manejar las distintas variantes es sumar las cifras de cada opción, dicho número corresponde al tipo de eliminación deseada.

Por ejemplo, `1+2+4+100=107` es eliminación por reducción rápida, a ambos lados de cada pivote, por filas (por tanto, la doble reducción significa que hay ceros por encima y por debajo de cada pivote, y de atrás hacia delante (como es por filas, aquí significa de abajo a arriba). Veámoslo:



```python
B.elim(107, rep=True)
```

Pero `2+10=12` es eliminación por doble reducción por columnas (es decir, ceros a derecha e izquierda de cada pivote), normalizando los pivotes para que sean unos:



```python
B.elim(12, rep=True)
```

Por tanto, la eliminación Gauss-Jordan, que además escalona, será `2+10+20=32`



```python
B.elim(32, rep=True)
```

Aunque el procedimiento `elim()` tiene una gran versatilidad, normalmente la eliminación se realiza para obtener un tipo concreto de matriz. Por ello, hay programados ciertos atajos. 

En los casos de eliminación por columnas, lo habitual es usar alguna de las siguientes:



```python
display(B.K()) # realiza los pasos conducentes a obtener una forma pre-escalonada por columnas
display(B.L()) # realiza los pasos conducentes a obtener una forma escalonada por columnas
display(B.R()) # realiza los pasos conducentes a obtener una forma escalonada-reducida por columnas
```

En los casos de eliminación por filas, lo habitual es usar alguna de las siguientes:



```python
U = display(B.U()) # realiza los pasos conducentes a obtener una forma escalonada por filas
UR = display(B.UR()) # realiza los pasos conducentes a obtener una forma escalonada-reducida por filas
```

Otras formas de obtener las tres primeras formas pre-escalonadas por columnas son:



```python
K = Elim(B)   # forma preescalonada por columnas (eliminación de izquierda a derecha). Equivalente a B.k()
L = ElimG(B)  # forma escalonada por columnas (eliminación gaussiana). Equivalente a B.L()
R = ElimGJ(B) # forma escalonada reducida por columnas (eliminación Gauss-Jordan). Equivalente a B.R()
Sistema([K,L,R])
```

Y las equivalentes por filas son:



```python
KF = ElimF(B)   # forma preescalonada por filas (eliminación de arriba a abajo)
U = ElimGF(B)  # forma escalonada por filas (eliminación gaussiana). Equivalente a B.U()
RF =ElimGJF(B) # forma escalonada reducida por filas (eliminación Gauss-Jordan). Equivalente a B.UR()
Sistema([KF,U,RF])
```

## Transformaciones de las filas



*Primero, lea la [sección 4.A](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.alph3.4.Alph1) del libro.*

Cuando las transformaciones elementales actúan por la izquierda de la matriz, realizan operaciones con las filas.

Por ejemplo, podemos intercambiar las dos primeras filas `C` del siguiente modo:



```python
display(B)
T( {1,3} ) & B.copy()
```

(*Recuerde que si aplica transformaciones elementales directamente sobre una matriz, modifica la matriz. Si quiere mantener la matriz original sin cambios, aplique las transformaciones sobre una copia*).

Tenga en cuenta que cuando actúan varias transformaciones por la izquierda, las primeras en actuar son las más próximas a la matriz; así, tanto



```python
T({1,2}) & T((2,2)) & Matrix(A)
```

como



```python
T([ {1,2}, (2,2) ]) & Matrix(A)
```

son equivalentes a (y fíjese en el uso de los paréntesis):



```python
T({1,2}) & ( T((2,2)) & Matrix(A) )
```

es decir, en los tres casos, primero se multiplica por 2 la segunda fila (operación dentro del paréntesis en el último ejemplo, pero la más próxima a la matriz en los dos primeros) y luego se intercambian las dos primeras filas.



### Producto de una matriz elemental por la izquierda de una matriz.



Tal como ocurre con las transformaciones de las columnas, podemos transformar las filas de una matriz al transformar las filas de la matriz identidad y luego multiplicar la matriz resultante por la matriz que queremos transformar:



```python
T((40,3,2)) & B.copy()
```

```python
matrizElemental =  T((40,3,2)) & I(3)
matrizElemental
```

```python
matrizElemental * B
```

Fíjese en que la notación no funciona de igual manera cuando la transformación elemental se aplica a las filas de la identidad. En el ejemplo, el `40` no está en la tercera fila, segunda columna. Si se fija, la matriz elemental que hemos obtenido al transformar las filas de $\Mat{I}$ es la transpuesta de la matriz elemental que obtendríamos si transformamos las columnas de la identidad. Es decir, al aplicar la transformación elemental por la izquierda, transponemos la matriz elemental que obtendríamos si aplicáramos la transformación por la derecha de $\Mat{I}$.



### Transformaciones de las filas y de las columnas



Si tenemos transformaciones actuando a la izquierda y a la derecha, operamos tanto con las filas como con las columnas.



```python
T( {1,2} ) & T( (2,2) ) & Matrix(A) & T( [(-3,2,3), (10,1), {2,3}] )
```

De hecho, el procedimiento `elim` permite encadenar distintos tipos de transformaciones. Por ejemplo, el siguiente código aplica la eliminación por filas (de arriba a abajo). Seguidamente, la eliminación por columnas (de izquierda a derecha). Luego normaliza los pivotes y finalmente escalona la matriz.



```python
B.elim(4).elim(0).elim(10).elim(20,rep=True)
```

## Decidiendo las operaciones por uno mismo



Imaginemos la siguiente matriz



```python
D = Matrix([[0,7,1],[-4,21,3]])
D
```

Si pedimos a `NAcAL` que pre-escalone la matriz, hará lo siguiente:



```python
ElimG(D, True)
```

&#x2026; que no es muy elegante, pues la librería sigue unas reglas fijas, aunque conduzcan a realizar más operaciones de las necesarias.

En este ejemplo, dado que el pivote de la última columna es un `1`, es mucho más sencillo:

1.  intercambiar la primera columna por la última,
2.  restar siete veces dicha columna a la segunda, y como se anulará, intercambiar la segunda columna con la tercera.

Es decir, aplicar la siguiente secuencia de transformaciones sobre las columnas:



```python
TrC = [ T( {1,3} ),
        T( [(-7,1,2), {2,3}] ) ]
```

Para verlo, vamos a usar `dispElim(Matriz, pasos)`, donde `pasos` es una lista de dos componentes. 

-   El primero corresponde a las transformaciones de las filas (en este ejemplo será una lista vacía porque no hay ninguna transformación que opere sobre las filas).
-   El segundo corresponde a las transformaciones a las columnas.

Estos componentes pueden ser: 

-   una lista formada por objetos de la clase `T`, en cuyo caso todas las transformaciones "empaquetadas" en cada `T` son mostradas como un paso en la sucesión de cambios de la matriz.
-   Un objeto de la clase `T`, en cuyo caso veremos la aplicación de las transformaciones elementales e intercambios de uno en uno.

Como `TrC` es una lista de objetos `T`, veremos la eliminación en los pasos 1. y 2. descritos más arriba:



```python
dispElim( D, [[], TrC]  )
```

Si el componente de pasos consiste en un único objeto de la clase `T`, entonces veremos cómo cambia la matriz con cada operación, de una en una:



```python
operacionesColumnas = T( [ {1,3}, (-7,1,2), {2,3} ] )

dispElim( D, [[], operacionesColumnas]  )
```

## Transposición de una secuencia de transformaciones elementales



Diremos que transponemos una secuencia de transformaciones al invertir el orden de las transformaciones en la secuencia. 
Al igual que con las matrices, transponemos con el operador `~` actuando por la izquierda de `T`.



```python
h=T([ (-2, 1, 2), (-3, 1, 3), (-4, 1, 4), (-5, 1, 5),  (-2, 2, 3), (4, 2, 4), (5, 2, 5) ])
h
```

```python
~h
```

Fíjese en que el orden de las operaciones se ha invertido tras la transposición.

Considere la matriz obtenida tras aplicar una sucesión de transformaciones sobre las *columnas de la matriz identidad* de orden $n$, obtenemos una matriz. 
Pues bien, la transpuesta de dicha sucesión aplicada sobre las *filas de la matriz identidad* de orden $n$ será la transpuesta de la anterior matriz.



```python
I(5) & h
```

```python
(~h) & I(5)
```

Fíjese que aplicar una sucesión de transformaciones es como <u>multiplicar</u> por una matriz, de manera que si aplicamos sobre las columnas de $\Mat{B}$



```python
B
```

una sucesión de transformaciones



```python
Matrix(B) & h
```

la transpuesta de dicha matriz resultante



```python
~(Matrix(B)&h)
```

se puede calcular con una operación que recuerda a la transpuesta del producto&#x2026; pues es la transpuesta de la transformación actuando sobre la transpuesta de `B`



```python
~h & ~Matrix(B)
```
