## Vector Addition



Generate the vectors $\Vect{a}$ and $\Vect{b}$:

-   The vector $\Vect{a}$ is formed by the list of numbers: 1, 4, 9, 2.
-   The vector $\Vect{b}$ is formed by the list of numbers: 1, 2, 1, 0.



```python
a = Vector([1, 4, 9, 2])
b = Vector([1, 2, 1, 0])
```

Add both vectors:



```python
a + b
```

Then, obtain the third component of the sum of both: $\quad \elemR{(\Vect{a}+\Vect{b})}{3}$.



```python
(a + b)|3
```

Now add the third component of $\Vect{a}$ with the third of $\Vect{b}$: $\quad \eleVRPE{a}{3}+\eleVRPE{b}{3}$.   
Verify that you get the same result as in the previous cell. 
What is this property called? (*you will find the name of the property in the [book](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#section.1.2)*)



```python
(a|3) + (b|3)
```

### Order of Operations and the Need for Parentheses



Now try removing the parentheses from the previous expression and run the code
(*remove the hashtag `#` and run the following code to see it*).



```python
# a|3 + b|3
```

Python will throw an error because it reads and executes operations sequentially from left to right: therefore, when it reaches the symbol `+`, it tries to add vector $\Vect{b}$ to the third component of $\Vect{a}$ (the number `9` you have already calculated), but the addition of numbers and vectors is not defined, resulting in an error. Thus, it is necessary to use parentheses to indicate the correct order of operations. This allows you to calculate first what is inside the parentheses and then perform the addition of the two results.

Note that when we humans read $\;\eleVR{a}{3}+\eleVR{b}{3},\;$ we mentally interpret it as $\;\eleVRPE{a}{3}+\eleVRPE{b}{3};\;$, but Python requires you to clearly indicate the order of operations with parentheses.

When expressions are complex, even humans need parentheses for proper interpretation. Whenever you write a mathematical expression and have doubts, it is better to use parentheses. Remember that, when working with Python, operations are executed left to right, and therefore, parentheses are usually necessary.
