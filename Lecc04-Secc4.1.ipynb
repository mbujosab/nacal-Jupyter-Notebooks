## Transformaciones elementales



Consideraremos dos tipos de transformaciones elementales que denotaremos con tuplas. 
Para indicar que una tupla en particular es una abreviatura de una transformación elemental, la escribiremos como argumento de la clase `T`.

-   **Tipo I:** Las transformaciones elementales del primer tipo suman $a$ veces el vector $i$-ésimo al vector $j$-ésimo:
    $$
           \underset{\left[\left(a\right)\mathbf{i}+\mathbf{j} \right]}{\mathbf{\tau}}
           $$
    Denotaremos estas transformaciones con una tupla de tres componentes: `T( (a, i, j) )`.
    
    Por ejemplo, la abreviatura de la transformación elemental que suma 3 veces el segundo vector al primer vector se escribe



```python
T( (3, 2, 1) )
```

Fíjese que el índice del vector modificado aparece en último lugar; que a dicho vector se le suma un múltiplo de otro vector, cuyo índice aparece en segundo lugar; y que el coeficiente por el que se multiplica aparece en primer lugar.

-   **Tipo II:** Las transformaciones elementales del segundo tipo multiplican el vector $j$-ésimo por un número $a$.
    $$
           \underset{\left[\left(a\right)\mathbf{j}\right]}{\mathbf{\tau}}
           $$
    Denotaremos estas transformaciones con una tupla de dos componentes: `T( (a, j) )`.
    
    Por ejemplo, la transformación elemental que multiplica por 5 al segundo vector se escribe



```python
T( (5, 2) )
```

Nuevamente, el índice del vector modificado aparece en último lugar, y el coeficiente por el que se multiplica aparece en primer lugar.



## Transformaciones elementales de las filas o las columnas de una matriz



### Transformaciones de las columnas



Cuando aplicamos una transformación elemental por la derecha a una matriz, modificamos sus columnas. La actuación de una transformación elemental se indica con el símbolo `&`

Por ejemplo, considere la matriz $\Mat{A}$



```python
A=Matrix([Vector((1,2)), Vector((0,3)), Vector((-2,2))])
A
```

Vamos a sumar el doble (2) de la primera columna (1) de $\Mat{A}$ a su tercera columna (3).



```python
A & T( (2,1,3) )
```

Cuando aplicamos una transformación elemental sobre una matriz, modificamos la matriz. Así que ahora la matriz $\Mat{A}$ ha cambiado:



```python
A
```

Si queremos aplicar transformaciones sobre una matriz y no perder la matriz original, necesitamos hacer una copia. Por ejemplo, definamos la siguiente matriz $\Mat{A}$



```python
x = Vector((1, 1, -2)); y = Vector((1, -2, 1)); z = Vector((-2, 1, 1))
A = Matrix([x,y,z])
A
```

y hagamos una copia (con `Matrix(A)`) a la que aplicaremos un par de transformaciones



```python
B = Matrix(A)
B & T( (2,1,3) ) & T( (5, 2) )
```

Ahora $\Mat{B}$ ha cambiado, pero $\Mat{A}$ se mantiene.



```python
A
```

### Sucesión de transformaciones elementales



*Léase la [sección 4.1.5](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.4.1.5) del libro.*

Es posible generar una sucesión de transformaciones elementales:
$$
\underset{\begin{subarray}{c} \left[\left(a\right)\mathbf {j}\right]\\\left[\left(b\right)\mathbf{p}+\mathbf{q} \right]\end
{subarray}}{\mathbf{\tau}}
$$
Para ello, podemos usar como argumento de la clase `T` <u>una lista de abreviaturas</u>. 
Por ejemplo:



```python
T( [(3, 2, 1), (5, 2)] )
```

#### Composición de transformaciones elementales



Pero, además, si aplicamos una transformación elemental a otra transformación elemental, logramos una transformación que es la composición de ambas. Por ejemplo



```python
z = T( (2,1,3) ) & T( (5, 2) )
z
```

Si ahora aplicamos `z` a una copia de `A`, obtendremos el mismo resultado que obtuvimos más arriba



```python
C = Matrix(A)
C & z
```

Los objetos de la clase `T` poseen un atributo `.abreviaturas`, donde se almacena el listado de abreviaturas. Por ejemplo, en el atributo `.abreviaturas` de `z` aparece el listado con las abreviaturas de las transformaciones aplicadas más arriba:



```python
z.abreviaturas
```

En este ejemplo, el atributo `.abreviaturas` consiste en una lista de tuplas (con dos o tres elementos), pero más abajo veremos que la lista `.abreviaturas` también puede contener conjuntos de dos índices `{i,j}` (para denotar intercambios).



##### Instanciando la clase `T` con una lista de abreviaturas (o una lista de listas de abreviaturas u objetos de la clase `T`)



Ya hemos visto que es posible instanciar `T` usando como argumento una sola abreviatura, pero también con una lista de abreviaturas. 
Por ejemplo:



```python
T( [(-3,1,2), (10,1), (5,6,3), (-9,6)] )
```

Por defecto, la lista de operaciones aparece en formato vertical, pero podemos indicar que queremos ver la lista en horizontal:



```python
T( [(-3,1,2), (10,1), (5,6,3), (-9,6)], rpr='h' )
```

También podemos usar como argumento:

-   una lista de listas de abreviaturas
-   otro objeto de la clase `T`
-   una lista de objetos de la clase `T`
-   una lista de listas de abreviaturas y de listas de objetos `T`

Ejemplo que usa una lista de listas



```python
Tr = T( [ [(-3,1,2), (10,1)], [(5,6,3), (-9,6)]], rpr='h' )
Tr
```

Ejemplo que usa un objeto de la clase `T`



```python
T( Tr, rpr='h' )
```

Ejemplo que usa una lista de objetos de la clase `T`



```python
TrC = [ T( [(-3,1,2), (10,1)] ),  T( [(5,6,3), (-9,6)] ) ]
T( TrC, rpr='h' )
```

Ejemplo que usa una lista de abreviaturas, listas de abreviaturas y de listas de objetos `T`



```python
T( [ (-3,1,2),  [T( [(10,1)] )],  [(5,6,3), (-9,6)] ], rpr='h'  )
```

El resultado es el mismo.



#### Intercambios



*Léase la [sección 4.1.6](https://mbujosab.github.io/CursoDeAlgebraLineal/libro.pdf#subsection.4.1.6) del libro.*

Hay múltiples combinaciones de transformaciones elementales que logran intercambiar de posición dos vectores. La siguiente es un ejemplo de secuencia que intercambia los vectores primero y tercero:



```python
display(A)
intercambio = T( [ (-1,3), (-1,3,1), (1,1,3), (-1,3,1) ] )
Matrix(A) & intercambio
```

Podemos ver los pasos con ayuda de `dispElim()` (más adelante veremos el funcionamiento de `dispElim()`):



```python
dispElim( A, [[], intercambio] )
```

Afortunadamente, hay una forma mucho más sencilla de realizar un intercambio.
Con un conjunto compuesto por dos índices indicamos el intercambio de los vectores correspondientes a dichos índices. 
Por ejemplo, `T( {s, t} )` intercambia los vectores $s$-ésimo y $t$-ésimo:
$$
\underset{\left[\mathbf{s}\rightleftharpoons\mathbf{t}\right]}{\mathbf{\tau}}
$$
Así, para intercambiar las columnas primera y tercera de una copia de $\Mat{A}$, basta con escribir



```python
Matrix(A) & T( {1,3} )
```

Si el conjunto solo contiene un índice, el intercambio no hace nada:



```python
Matrix(A) & T( {1,1} )
```

## Matrices elementales, matrices intercambio y matrices permutación



La aplicación de una *única* transformación elemental sobre la matriz identidad da como resultado una matriz elemental.

-   **Matrices elementales Tipo I:** Veamos cómo la aplicación de la transformación `T(-5,1,3)` sobre las *columnas* de la matriz identidad de orden 4 nos da una matriz parecida a la identidad, salvo porque aparece un $-5$ en la posición $(1,3)$.



```python
I(4) & T( (-5,1,3) )
```

Pero la misma transformación elemental aplicada sobre las *filas* de la matriz identidad nos da otra matriz elemental distinta: la versión transpuesta de la matriz elemental de más arriba:



```python
T( (-5,1,3) ) & I(4)
```

Por tanto, las transformaciones elementales de *Tipo I* están asociadas a matrices elementales distintas en función de si actúan sobre las columnas (por la derecha) o sobre las filas (por la izquierda). Fíjese que una matriz elemental es la transpuesta de la otra.

-   **Matrices elementales Tipo II:** Y ahora veamos cómo la aplicación de la transformación `T(100,2)` sobre las *columnas* de la matriz identidad de orden 3 nos da una matriz parecida a la identidad, salvo porque aparece un $100$ en la segunda posición de la diagonal, es decir, en la posición $(2,2)$.



```python
I(3) & T( (100,2) )
```

Las matrices elementales *Tipo II* son simétricas; así que obtenemos la misma matriz elemental cuando actuamos por la izquierda sobre la matriz identidad.



```python
T( (100,2) ) & I(3)
```

Tenga en cuenta que la aplicación de <u>una sucesión de transformaciones elementales *NO* suele ser una matriz elemental</u>.



```python
I(3) & T( (2,1,3) ) & T( (5, 2) ) & T( (-5,3,1) ) & T ( (2,1,2) )
```

-   **Matrices intercambio:** La aplicación de  <u>un único intercambio</u> entre las columnas (o entre las filas) de la matriz identidad da como resultado una matriz intercambio.



```python
I(4) & T( {2,4} )
```

```python
T( {2,4} ) & I(4)
```

-   **Matrices permutación:** La aplicación de una <u>sucesión de intercambios</u> entre las columnas (y/o entre las filas) de la matriz identidad da como resultado una reordenación de las columnas de la matriz identidad que denominamos matriz permutación.



```python
I(4) & T({3,1}) & T({2,3}) & T({1,2}) & T({3,4})
```

que es como una matriz identidad, pero con sus filas o columnas reordenadas.

Los intercambios se pueden realizar tanto por la izquierda (filas) como por la derecha (columnas).



```python
T( [ {1,2}, {3,4} ] ) & I(4) & T({3,1}) & T({2,3})
```

### Producto de una matriz por una matriz elemental



Compruebe con ejemplos que aplicar una transformación elemental sobre las columnas de una matriz es equivalente a multiplicar dicha matriz por la correspondiente matriz elemental; es decir, que 
$$
{\Mat{A}}_{\mathbf{\tau}}=\Mat{A}\big({\Mat{I}}_{\mathbf{\tau}}\big)
$$



```python
# Primero aplicamos la transformación sobre las columnas
Matrix(A) & T( (2,1,3) )
```

```python
# Y ahora multiplicamos por la derecha por la correspondiente matriz elemental
A * ( I(A.n) &  T((2,1,3)) )
```

Compruebe que también ocurre actuando por la izquierda para transformar las filas. Primero aplicamos la transformación sobre las filas.



```python
# Primero aplicamos la transformación sobre las filas
T( (2,1,3) ) & Matrix(A)
```

```python
# Y ahora multiplicamos por la izquierda por la correspondiente matriz elemental
( T((2,1,3)) & I(A.m) ) * A
```

### Las transformaciones son asociativas respecto al producto de matrices



Compruebe con ejemplos que en general
$$
{\big(\Mat{A}\Mat{B}\big)}_{\mathbf{\tau}}=\Mat{A}\big({\Mat{B}}_{\mathbf{\tau}}\big)
$$
Recuerde emplear una copia de las matrices originales si no quiere que éstas cambien.



```python
# Primero el lado izquierdo de la igualdad
( A * Matrix(B) ) & T((-5,2,1))
# Recuerde que si no queremos modificar las matrices originales, debemos actuar sobre copias de las mismas.
```

```python
# Ahora el lado derecho de la igualdad
A * ( Matrix(B) & T((-5,2,1)) )
# Recuerde que si no queremos modificar las matrices originales, debemos actuar sobre copias de las mismas.
```

Esto es cierto incluso cuando se aplica una sucesión de transformaciones elementales (tanto por la derecha como por la izquierda).

Veamos un ejemplo de transformación de las columnas:



```python
# Definimos una sucesión de transformaciones elementales
st = T([(2, 1, 3), (5, 2), (-5, 3, 1), (2, 1, 2), {2,3}])
st
```

```python
( A * Matrix(B) ) & st
```

```python
A * ( Matrix(B) & st )
```

Y ahora un ejemplo de transformación de las filas&#x2026;



```python
st & ( Matrix(A) *B )
```

```python
( st & Matrix(A) ) * B
```

## Las transformaciones elementales son operadores lineales



### Propiedad asociativa



Compruebe con ejemplos que en general
$$
{\big(\lambda\Mat{B}\big)}_{\mathbf{\tau}}=\lambda\big({\Mat{B}}_{\mathbf{\tau}}\big)
$$
Recuerde emplear una copia de las matrices originales para que éstas no cambien.



```python
# Primero el lado izquierdo de la igualdad
( 5 * Matrix(A) ) & st
```

```python
# Y luego el lado derecho
5 * ( Matrix(A) & st )
```

Compruebe, usando los dos recuadros de más abajo, la propiedad asociativa cuando se transforman las filas



### Propiedad distributiva



Compruebe con ejemplos que en general
$$
{\big(\Mat{A}+\Mat{B}\big)}_{\mathbf{\tau}}={\Mat{A}}_{\mathbf{\tau}} + {\Mat{B}}_{\mathbf{\tau}}
$$



```python
# Primero el lado izquierdo de la igualdad
( Matrix(A) + Matrix(B) ) & st
```

```python
# Y luego el lado derecho de la igualdad
( Matrix(A) & st ) + ( Matrix(B) & st )
```

Esto también es cierto con las transformaciones de las filas



```python
st & ( Matrix(A) + Matrix(B) )
```

```python
( st & Matrix(A) ) + ( st & Matrix(B) )
```
